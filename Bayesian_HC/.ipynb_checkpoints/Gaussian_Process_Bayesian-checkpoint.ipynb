# Nonparametric Bayesian Inference: Gaussian Processes (GPs)

## 0. Basic Framework of Nonparametric Bayesian Inference

We know that a statistical model is essentially a family (or measure) of distributions over a sample space ð’³:  
$$
\{ P_\theta : \theta \in \Theta \}
$$

In parametric statistics, we usually assume that Î˜ is finite-dimensional (e.g., $ \mathcal{N}(\theta, I_n), \theta \in \mathbb{R}^n $);  
whereas in nonparametric statistics, we usually assume that Î˜ is infinite-dimensional (e.g., the set of all continuous functions over $ \mathbb{R} $).

The core idea of Bayesian statistics is to treat the parameter itself as a random variable, and to assign a **prior distribution** $ \pi(\theta) $ over the parameter space.  
Using **Bayes' Theorem**, we compute or approximate the **posterior distribution** of the parameter based on observed data sampled from the sample space:

$$
p(\theta | X_1, \ldots, X_n) \propto \pi(\theta) \prod_{i=1}^n p(X_i | \theta)
$$

This posterior distribution is then used to perform inference.  
Unlike parametric Bayesian methods, nonparametric Bayesian inference requires us to define a proper prior over an infinite-dimensional parameter space.

---

> A function follows a Gaussian Process if, at any finite set of input points, the function values follow a multivariate Gaussian distribution.



## 1. From Bayesian Linear Regression to Gaussian Processes

Consider a Bayesian linear regression model:

$$ f(x) = w^T x, \quad \text{where} \quad w \sim \mathcal{N}(0, \Sigma_p) $$

Then:

$$ f(x) \sim \mathcal{N}(0, x^T \Sigma_p x) $$

For multiple input points:

$$ f = Xw, \quad X = \begin{bmatrix} x_1^T \\ x_2^T \\ \vdots \\ x_n^T \end{bmatrix} \Rightarrow f \sim \mathcal{N}(0, X \Sigma_p X^T) $$

This implies: function values at any finite set of inputs follow a multivariate Gaussian.

This is the core idea behind GPs.

```python
# Linear model: generate GP samples
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import multivariate_normal

X = np.linspace(-5, 5, 100).reshape(-1, 1)
Sigma_p = np.eye(X.shape[1])
cov = X @ Sigma_p @ X.T

samples = multivariate_normal(mean=np.zeros(X.shape[0]), cov=cov, size=3)

plt.plot(X, samples.T)
plt.title("Samples from Bayesian Linear Model")
plt.show()
```

## 2. Definition of a Gaussian Process

A Gaussian Process is defined by a mean and covariance function:

$$ f(x) \sim \mathcal{GP}(m(x), k(x, x')) $$

Where:
- $ m(x) = \mathbb{E}[f(x)] $ (mean function)
- $ k(x,x') = \mathbb{E}[(f(x) - m(x))(f(x') - m(x'))] $ (kernel)

Typically:
- Mean is assumed to be zero.
- Kernel can be linear, RBF, periodic, etc.

**Example: RBF kernel**

$$ k(x, x') = \sigma_f^2 \exp\left( -\frac{ \| x - x' \|^2 }{2\ell^2} \right) $$

```python
def rbf_kernel(X1, X2, length_scale=1.0, sigma_f=1.0):
    sqdist = np.sum(X1**2, 1).reshape(-1, 1) + np.sum(X2**2, 1) - 2 * X1 @ X2.T
    return sigma_f**2 * np.exp(-0.5 / length_scale**2 * sqdist)
```

## 3. Gaussian Process Regression

### 3.1 Problem Setup

Given:
- Training data $ X \in \mathbb{R}^{n \times d} $
- Observations: $ y_i = f(x_i) + \epsilon_i $, $ \epsilon_i \sim \mathcal{N}(0, \sigma_n^2) $

We want to predict $ f_* $ at test points $ X_* $.

### 3.2 Joint Distribution

$$
\begin{bmatrix}
\mathbf{y} \\\\
\mathbf{f}_*
\end{bmatrix}
\sim \mathcal{N}\left(0,
\begin{bmatrix}
K(X,X) + \sigma_n^2 I & K(X,X_*) \\\\
K(X_*,X) & K(X_*,X_*)
\end{bmatrix}
\right)
$$

### 3.3 Posterior Predictive Distribution

- Predictive Mean:

$$
\bar{f}_* = K(X_*, X)[K(X,X) + \\sigma_n^2 I]^{-1} y
$$

- Predictive Covariance:

$$
\text{Cov}(f_*) = K(X_*, X_*) - K(X_*, X)[K(X,X) + \\sigma_n^2 I]^{-1}K(X, X_*)
$$

```python
# GP posterior prediction
from numpy.linalg import inv

# Example training data
X = np.linspace(-5, 5, 20).reshape(-1, 1)
y = np.sin(X).ravel() + 0.1 * np.random.randn(20)

# Test inputs
X_test = np.linspace(-5, 5, 100).reshape(-1, 1)

# Kernel and noise parameters
sigma_n = 0.1
K = rbf_kernel(X, X)
K_s = rbf_kernel(X, X_test)
K_ss = rbf_kernel(X_test, X_test)
K_noise = K + sigma_n**2 * np.eye(len(X))

# Posterior mean and covariance
mu_s = K_s.T @ inv(K_noise) @ y
cov_s = K_ss - K_s.T @ inv(K_noise) @ K_s

# Plot results
import matplotlib.pyplot as plt

std_s = np.sqrt(np.diag(cov_s))
plt.figure(figsize=(10, 5))
plt.plot(X, y, 'ro', label="Training data")
plt.plot(X_test, mu_s, 'b-', label="Predictive mean")
plt.fill_between(
    X_test.ravel(),
    mu_s - 1.96 * std_s,
    mu_s + 1.96 * std_s,
    color="blue",
    alpha=0.2,
    label="95% confidence interval"
)
plt.legend()
plt.title("Gaussian Process Regression")
plt.show()
```

## 4. Kernel Choices and Hyperparameter Tuning

Different kernels model different function behaviors:

- **Linear**: linear trends
- **RBF**: smooth, non-linear functions
- **Periodic**: repeating patterns
- **MatÃ©rn**: flexible smoothness

### Hyperparameter Learning

Estimate kernel parameters (e.g., $ \ell, \sigma_f $) by maximizing the log marginal likelihood:

$$
\log p(y | X) = -\frac{1}{2} y^T (K + \sigma_n^2 I)^{-1} y - \frac{1}{2} \\log |K + \sigma_n^2 I| - \frac{n}{2} \log 2\pi
$$

You can use libraries like `scikit-learn`, `GPy`, or `GPflow` for automated hyperparameter optimization.

## 5. Conclusion

- Gaussian Processes define a distribution over functions.
- They are fully specified by a mean and covariance function.
- The posterior distribution allows principled predictions with uncertainty.
- Computational complexity is $ O(n^3) $, limiting scalability.

> With expressive kernels and a Bayesian framework, GPs are powerful tools for regression and beyond.

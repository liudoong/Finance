## 1. Basic Framework

Let $(\Omega, \mathcal{F}, \mathbb{P})$ be a complete probability space, where:
- $\Omega$ is the sample space
- $\mathcal{F}$ is a $\sigma$-algebra of observable
- $\mathbb{P}$ is the risk-neutral probability measure

Let $\mathbb{F} = (\mathcal{F}_t)_{t \in [0,T]}$ be a filtration satisfying the usual conditions, representing the flow of information over time.


The underlying asset price is modeled as a stochastic process $(S_t)_{t \in [0,T]}$ adapted to the filtration $\mathbb{F}$. We assume $S_t$ is a positive semimartingale, which allows for both continuous evolution and jumps.

**Definition:  Underlying Asset Proces** 
$$S_t: [0,T] \times \Omega \to \mathbb{R}^+ \quad s.t.: \quad \forall \quad t \in [0,T] \quad \land \quad S_t \subset \mathcal{F}_t-measurable. $$



An exotic equity derivative is characterized by its payoff function $\Phi: \mathcal{P}([0,T] \times \mathbb{R}^+) \to \mathbb{R}^+$, where $\mathcal{P}([0,T] \times \mathbb{R}^+)$ denotes the path space.

**Definition: Path Functional**

Let $X: \Omega \to C([0,T], \mathbb{R}^+)$ be the path mapping defined by $X(\omega) = (S_t(\omega))_{t \in [0,T]}$. The payoff is a measurable function:
$$\Phi: C([0,T], \mathbb{R}^+) \to \mathbb{R}^+$$

The terminal payoff at time $T$ is given by:
$$V_T = \Phi(X).$$


We define several most frequently used fundamental path functionals that will be used to construct exotic payoffs.

**Definition: Running Maximum Functional**
$$M_t = \max_{0 \leq \tau \leq t} S_\tau = \sup\{S_\tau : \tau \in [0,t]\}$$

**Definition: Running Minimum Runctional**
$$m_t = \min_{0 \leq \tau \leq t} S_\tau = \inf\{S_\tau : \tau \in [0,t]\}$$

**Definition: Indicator Functional**
$$ \forall \quad Borel \quad set \quad \mathfrak{A} \subset \mathbb{R}^+ \qquad
s.t.: \quad \mathbf{1}_\mathfrak{A}(S_t) = \begin{cases}
1 & \text{if } S_t \in \mathfrak{A} \\
0 & \text{otherwise}
\end{cases}$$

**Definition: Barrier Hit Indicator**

$$\forall \quad barrier \quad \mathfrak{B}>0 \quad \tau_\mathfrak{B} = \inf\{t \in [0,T] : S_t \leq \mathfrak{B}\} \quad \text{(down-barrier)}$$
$$\mathbf{1}_{\{\tau_\mathfrak{B} \leq T\}} = \begin{cases}
1 & \text{if the barrier is hit by time } T \\
0 & \text{otherwise}
\end{cases}$$

and on the opposite side:

$$\forall \quad barrier \quad \mathfrak{B}>0 \quad \tau_\mathfrak{B} = \sup\{t \in [0,T] : S_t \geq \mathfrak{B}\} \quad \text{(up-barrier)}$$
$$\mathbf{1}_{\{\tau_\mathfrak{B} \leq T\}} = \begin{cases}
1 & \text{if the barrier is hit by time } T \\
0 & \text{otherwise}
\end{cases}$$

If we assume 252 trading days per year, for higher moments underlyings, the realized variance and volatility are defined as following:

**Definition: Realized Variance**

$$\forall \quad partition \quad 0 = t_0 < t_1 < \cdots < t_N = T \quad s.t.: \quad \sigma_R^2 = \frac{252}{N} \sum_{i=1}^N \left[\ln\left(\frac{S_{t_i}}{S_{t_{i-1}}}\right)\right]^2$$

**Definition: Volatility Barrier Indicator**

$$\forall \quad Barrier \quad \mathfrak{B}_\sigma > 0 \quad s.t.: \quad \mathbf{1}_{\{\sigma_R^2 > \mathfrak{B}_\sigma\}} = \begin{cases}
1 & \text{if realized variance exceeds barrier} \\
0 & \text{otherwise}
\end{cases}$$

## 2. Multi-Asset Framework

For derivatives involving multiple assets, let $\mathbf{S}_t = (S_t^1, S_t^2, \dots, S_t^n)$ be an $n$-dimensional semimartingale.

**Definition: (Weighted Basket)**

Given weights $\mathbf{w} = (w_1, \dots, w_n)$ with $\sum_{i=1}^n w_i = 1$, the basket value is:
$$B_t = \sum_{i=1}^n w_i S_t^i$$

**Definition: (Extremu Functionals)**

For multi-asset paths:
- Best performer: $\max(\mathbf{S}_T) = \max\{S_T^1, S_T^2, \dots, S_T^n\}$
- Worst performer: $\min(\mathbf{S}_T) = \min\{S_T^1, S_T^2, \dots, S_T^n\}$

## 3. General Derivative Valuation

The no-arbitrage price of the derivative at time $t$ is given by:

**Theorem: (Risk-Neutral Pricing)**
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\Phi(X) \middle| \mathcal{F}_t\right]$$

where $r$ is the risk-free rate and $\mathbb{Q}$ is the risk-neutral measure.

**Proof:** This follows from the Fundamental Theorem of Asset Pricing and the martingale representation theorem. The discounted price process $e^{-rt}V_t$ must be a $\mathbb{Q}$-martingale, leading to the conditional expectation representation.

## 4. Specific Payoff and Potential Future Exposure (PFE)

### 4.1 Lookback Put:

**Payoff at Expiry:**
$$V_T = \max\left(M_T - S_T, 0\right)$$
where $M_T = \sup\{S_\tau : \tau \in [0,T]\}$ is the running maximum.

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where $V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\max\left(M_T - S_T, 0\right) \middle| \mathcal{F}_t\right]$

### 4.2 Reset Put: Payoff and Potential Future Exposure

**Payoff at Expiry:**
$$V_T = \max(K_{\text{reset}} - S_T, 0)$$
where the reset strike is defined as:
$$K_{\text{reset}} = \begin{cases}
S_{t_r} & \text{if } S_{t_r} < K \\
K & \text{otherwise}
\end{cases}$$
for reset time $t_r \in [0,T]$ and initial strike $K$.

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\max(K_{\text{reset}} - S_T, 0) \middle| \mathcal{F}_t\right]$$
and $K_{\text{reset}}$ is $\mathcal{F}_{t_r}$-measurable.

### 4.3 Put Knock-Out (Down-and-Out Put): Payoff and Potential Future Exposure

**Payoff at Expiry:**
$$V_T = \mathbf{1}_{\{\tau_B > T\}} \cdot \max(K - S_T, 0)$$
where $\tau_B = \inf\{t \in [0,T] : S_t \leq B\}$ is the first hitting time of barrier $B < S_0$.

Equivalently:
$$V_T = \begin{cases}
\max(K - S_T, 0) & \text{if } S_t > B \text{ for all } t \in [0,T] \\
0 & \text{if } \exists t \in [0,T] \text{ such that } S_t \leq B
\end{cases}$$

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\mathbf{1}_{\{\tau_B > T\}} \cdot \max(K - S_T, 0) \middle| \mathcal{F}_t\right]$$
and $\tau_B$ is a stopping time adapted to the filtration $\mathbb{F}$.

### 4.4 Call Up-and-Out: Payoff and Potential Future Exposure

**Payoff at Expiry:**
$$V_T = \mathbf{1}_{\{\tau_B < T\}^c} \cdot \max(S_T - K, 0)$$
where $\tau_B = \inf\{t \in [0,T] : S_t \geq B\}$ is the first hitting time of barrier $B > S_0$, and $\{\tau_B < T\}^c$ denotes the complement event (barrier not hit).

Equivalently:
$$V_T = \begin{cases}
\max(S_T - K, 0) & \text{if } S_t < B \text{ for all } t \in [0,T] \\
0 & \text{if } \exists t \in [0,T] \text{ such that } S_t \geq B
\end{cases}$$

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\mathbf{1}_{\{\tau_B > T\}} \cdot \max(S_T - K, 0) \middle| \mathcal{F}_t\right]$$
and $\tau_B$ is a stopping time adapted to the filtration $\mathbb{F}$.

### 4.5 Put Knock-In (Down-and-In Put): Payoff and Potential Future Exposure

**Payoff at Expiry:**
$$V_T = \mathbf{1}_{\{\tau_B \leq T\}} \cdot \max(K - S_T, 0)$$
where $\tau_B = \inf\{t \in [0,T] : S_t \leq B\}$ is the first hitting time of barrier $B < S_0$.

Equivalently:
$$V_T = \begin{cases}
\max(K - S_T, 0) & \text{if } \exists t \in [0,T] \text{ such that } S_t \leq B \\
0 & \text{if } S_t > B \text{ for all } t \in [0,T]
\end{cases}$$

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\mathbf{1}_{\{\tau_B \leq T\}} \cdot \max(K - S_T, 0) \middle| \mathcal{F}_t\right]$$
and $\tau_B$ is a stopping time adapted to the filtration $\mathbb{F}$.

**Note:** The knock-in and knock-out puts satisfy the parity relationship:
$$V_t^{\text{Down-and-In}} + V_t^{\text{Down-and-Out}} = V_t^{\text{Vanilla Put}}$$
for all $t \in [0,T]$, which follows from the identity $\mathbf{1}_{\{\tau_B \leq T\}} + \mathbf{1}_{\{\tau_B > T\}} = 1$.

### 4.7 Conditional Payout Option: Payoff and Potential Future Exposure

**Payoff at Expiry:**
$$V_T = \mathbf{1}_{\{Y_T \in A\}} \cdot \max(S_T - K, 0)$$
where:
- $S_T$ is the price of the primary underlying asset
- $Y_T$ is the price of the conditional underlying asset  
- $A \subset \mathbb{R}^+$ is the Borel set defining the payout condition (e.g., $A = (100, \infty)$ for $TLT_T > 100$)

Equivalently:
$$V_T = \begin{cases}
\max(S_T - K, 0) & \text{if } Y_T \in A \\
0 & \text{otherwise}
\end{cases}$$

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\mathbf{1}_{\{Y_T \in A\}} \cdot \max(S_T - K, 0) \middle| \mathcal{F}_t\right]$$

**Note:** This option introduces dependence on the joint distribution of $(S_t, Y_t)$, requiring modeling of the correlation structure between the primary and conditional underlyings. The indicator function $\mathbf{1}_{\{Y_T \in A\}}$ makes the payoff path-dependent on the conditional asset's terminal value.

### 4.8 Volatility Knock-Out Option: Payoff and Potential Future Exposure

**Payoff at Expiry:**
$$V_T = \mathbf{1}_{\{\sigma_{realized}(t) \leq B \ \forall t \in \mathcal{T}\}} \cdot \max(S_T - K, 0)$$
where:
- $\mathcal{T} \subset [0,T]$ is the set of observation times
- $\sigma_{realized}(t)$ is the realized volatility measured over a specified lookback window ending at time $t$
- $B$ is the volatility barrier level

Equivalently:
$$V_T = \begin{cases}
\max(S_T - K, 0) & \text{if } \sigma_{realized}(t) \leq B \text{ for all } t \in \mathcal{T} \\
0 & \text{if } \exists t \in \mathcal{T} \text{ such that } \sigma_{realized}(t) > B
\end{cases}$$

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\mathbf{1}_{\{\sigma_{realized}(\tau) \leq B \ \forall \tau \in \mathcal{T} \cap [0,T]\}} \cdot \max(S_T - K, 0) \middle| \mathcal{F}_t\right]$$

**Note:** This option introduces path-dependence through the volatility process rather than the price process. The knockout condition depends on the historical volatility trajectory, making the option sensitive to the joint dynamics of the underlying price and its volatility. The indicator function $\mathbf{1}_{\{\sigma_{realized}(t) \leq B \ \forall t \in \mathcal{T}\}}$ creates a complex barrier condition in volatility space.

### 4.9 Variance Knock-Out Option: Payoff and Potential Future Exposure

**Payoff at Expiry:**
$$V_T = \mathbf{1}_{\{\sigma_R^2 \leq B\}} \cdot \max(S_T - K, 0)$$
where:
- $\sigma_R^2$ is the total realized variance over $[0,T]$
- $B$ is the variance barrier level

The realized variance is defined as:
$$\sigma_R^2 = \frac{252}{N} \sum_{i=1}^N \left[\ln\left(\frac{S_{t_i}}{S_{t_{i-1}}}\right)\right]^2$$
for a partition $0 = t_0 < t_1 < \cdots < t_N = T$ of the option's life.

Equivalently:
$$V_T = \begin{cases}
\max(S_T - K, 0) & \text{if } \sigma_R^2 \leq B \\
0 & \text{if } \sigma_R^2 > B
\end{cases}$$

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\mathbf{1}_{\{\sigma_R^2 \leq B\}} \cdot \max(S_T - K, 0) \middle| \mathcal{F}_t\right]$$

**Note:** Unlike the volatility knock-out which checks barrier conditions at multiple observation times, the variance knock-out depends only on the cumulative variance over the entire option lifespan. This creates a terminal barrier condition in variance space, making the option sensitive to the quadratic variation of the underlying process. The indicator $\mathbf{1}_{\{\sigma_R^2 \leq B\}}$ introduces dependence on the entire price path through its realized variance.

### 4.10 Appearing Call Spread: Payoff and Potential Future Exposure

**Payoff at Expiry:**
$$V_T = \max(S_T - K_1, 0) - \mathbf{1}_{\{\tau_B \leq T\}} \cdot \max(S_T - K_2, 0)$$
where:
- $K_1$ is the initial strike price (lower strike)
- $K_2$ is the upper strike that appears when triggered ($K_2 > K_1$)
- $\tau_B = \inf\{t \in [0,T] : S_t \geq B\}$ is the first hitting time of trigger level $B$ (typically $B > K_1$)

Equivalently:
$$V_T = \begin{cases}
\max(S_T - K_1, 0) - \max(S_T - K_2, 0) & \text{if } \exists t \in [0,T] \text{ such that } S_t \geq B \\
\max(S_T - K_1, 0) & \text{if } S_t < B \text{ for all } t \in [0,T]
\end{cases}$$

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\left(\max(S_T - K_1, 0) - \mathbf{1}_{\{\tau_B \leq T\}} \cdot \max(S_T - K_2, 0)\right) \middle| \mathcal{F}_t\right]$$

**Note:** This derivative exhibits conditional payoff structure transformation. Before the trigger event, it behaves as a vanilla call. After triggering, it becomes a bull call spread with capped upside of $(K_2 - K_1)$. The indicator function $\mathbf{1}_{\{\tau_B \leq T\}}$ creates path-dependence that alters the fundamental payoff profile, making the option sensitive to both the terminal price and the maximum price achieved during the option's life.

### 4.11 Appearing Put Spread: Payoff and Potential Future Exposure

**Payoff at Expiry:**
$$V_T = \max(K_1 - S_T, 0) - \mathbf{1}_{\{\tau_B \leq T\}} \cdot \max(K_2 - S_T, 0)$$
where:
- $K_1$ is the initial strike price (higher strike)
- $K_2$ is the lower strike that appears when triggered ($K_2 < K_1$)
- $\tau_B = \inf\{t \in [0,T] : S_t \leq B\}$ is the first hitting time of trigger level $B$ (typically $B < K_1$)

Equivalently:
$$V_T = \begin{cases}
\max(K_1 - S_T, 0) - \max(K_2 - S_T, 0) & \text{if } \exists t \in [0,T] \text{ such that } S_t \leq B \\
\max(K_1 - S_T, 0) & \text{if } S_t > B \text{ for all } t \in [0,T]
\end{cases}$$

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\left(\max(K_1 - S_T, 0) - \mathbf{1}_{\{\tau_B \leq T\}} \cdot \max(K_2 - S_T, 0)\right) \middle| \mathcal{F}_t\right]$$

**Note:** This derivative exhibits conditional payoff structure transformation similar to the appearing call spread but for put options. Before the trigger event, it behaves as a vanilla put. After triggering, it becomes a bear put spread with capped upside of $(K_1 - K_2)$. The indicator function $\mathbf{1}_{\{\tau_B \leq T\}}$ creates path-dependence that transforms the payoff profile, making the option sensitive to both the terminal price and the minimum price achieved during the option's life.


### 4.12 Fixed Strike Forward Volatility Agreement (FVA): Payoff and Potential Future Exposure

**Payoff at Settlement:**
$$V_T = N \cdot (\sigma_{IV} - K_{\sigma})$$
where:
- $N$ is the notional amount (in volatility points or currency per volatility point)
- $\sigma_{IV}$ is the realized implied volatility of the reference option observed at time $T$
- $K_{\sigma}$ is the fixed volatility strike agreed at inception

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)} \cdot N \cdot (\sigma_{IV} - K_{\sigma}) \middle| \mathcal{F}_t\right]$$

**Note:** Unlike other derivatives discussed, the FVA is a forward contract on implied volatility rather than a contingent claim on the underlying asset price. The payoff is linear in the realized implied volatility and can be both positive and negative, representing the bilateral nature of forward agreements. The key modeling challenge lies in characterizing the dynamics of the implied volatility process $\sigma_{IV}$, which depends on market expectations of future volatility rather than historical realized volatility.

### 4.13 Rainbow Options - Worst-of Options

#### 4.13.1 Worst-of Call Option

**Payoff at Expiry:**
$$V_T = \max\left( \min(S_T^1, S_T^2, \dots, S_T^n) - K, 0 \right)$$
where:
- $S_T^i$ is the price of asset $i$ at expiry
- $K$ is the strike price
- $\min(S_T^1, S_T^2, \dots, S_T^n)$ selects the worst-performing asset

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)} \max\left( \min(S_T^1, S_T^2, \dots, S_T^n) - K, 0 \right) \middle| \mathcal{F}_t\right]$$

#### 4.13.2 Worst-of Put Option

**Payoff at Expiry:**
$$V_T = \max\left( K - \max(S_T^1, S_T^2, \dots, S_T^n), 0 \right)$$
where $\max(S_T^1, S_T^2, \dots, S_T^n)$ selects the best-performing asset (worst for the put)

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)} \max\left( K - \max(S_T^1, S_T^2, \dots, S_T^n), 0 \right) \middle| \mathcal{F}_t\right]$$

**Note:** Worst-of options are highly sensitive to the correlation structure among the underlying assets. Lower correlation increases the dispersion of returns, making the worst-of call less valuable (as the minimum of the assets is likely lower) and the worst-of put more valuable (as the maximum of the assets is likely higher). These options require modeling the joint distribution of all $n$ assets and their dependence structure.

#### 4.14.2 Best-of Put Option

$$V_T = \max\left( K - \min(S_T^1, S_T^2, \dots, S_T^n), 0 \right)$$

### 4.15 Rainbow Options - Outperformance Options

#### 4.15.1 Spread Outperformance Option

**Payoff at Expiry:**
$$V_T = \max\left( \max(S_T^1, S_T^2, \dots, S_T^n) - \min(S_T^1, S_T^2, \dots, S_T^n) - K, 0 \right)$$
where:
- $S_T^i$ is the price of asset $i$ at expiry
- $K$ is the strike price (often set to 0)

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)} \max\left( \max(S_T^1, \dots, S_T^n) - \min(S_T^1, \dots, S_T^n) - K, 0 \right) \middle| \mathcal{F}_t\right]$$

#### 4.15.2 Two-Asset Outperformance Call

**Payoff at Expiry (Relative Performance):**
$$V_T = \max\left( \frac{S_T^A}{S_0^A} - \frac{S_T^B}{S_0^B}, 0 \right)$$

**Payoff at Expiry (Absolute Spread):**
$$V_T = \max\left( S_T^A - S_T^B - K, 0 \right)$$

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)} \max\left( S_T^A - S_T^B - K, 0 \right) \middle| \mathcal{F}_t\right]$$

#### 4.15.3 Relative Outperformance Option

**Payoff at Expiry:**
$$V_T = \max\left( \frac{S_T^A - S_T^B}{S_T^B}, 0 \right)$$

**Alternative Formulation with Strike:**
$$V_T = \max\left( \frac{S_T^A}{S_T^B} - K, 0 \right)$$

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)} \max\left( \frac{S_T^A}{S_T^B} - K, 0 \right) \middle| \mathcal{F}_t\right]$$

**Note:** Outperformance options are fundamentally correlation products that pay based on the relative performance between assets. The spread outperformance option benefits from high dispersion (low correlation), while the two-asset and relative outperformance options express directional views on one asset's performance relative to another. These options require modeling the joint dynamics and co-dependence of multiple assets, with the relative performance versions introducing additional complexity through ratio-based payoffs.

### 4.16 Basket Call Option

**Payoff at Expiry:**
$$V_T = \max\left( \sum_{i=1}^n w_i S_T^i - K, 0 \right)$$
where:
- $S_T^i$ is the price of asset $i$ at expiry
- $w_i$ is the weight of asset $i$ in the basket ($\sum_{i=1}^n w_i = 1$)
- $K$ is the strike price
- $n$ is the number of underlying assets in the basket

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)} \max\left( \sum_{i=1}^n w_i S_T^i - K, 0 \right) \middle| \mathcal{F}_t\right]$$

**Note:** Basket call options provide leveraged exposure to the overall performance of a diversified portfolio. The option's value is heavily influenced by the correlation structure among the basket components - lower correlation reduces the basket's overall volatility due to diversification effects, making the option less valuable than a portfolio of individual calls. The weighted sum $\sum_{i=1}^n w_i S_T^i$ represents the basket value, and the option pays off when this diversified portfolio appreciates above the strike level.

### 4.17 Basket Put Option

**Payoff at Expiry:**
$$V_T = \max\left( K - \sum_{i=1}^n w_i S_T^i, 0 \right)$$
where:
- $S_T^i$ is the price of asset $i$ at expiry
- $w_i$ is the weight of asset $i$ in the basket ($\sum_{i=1}^n w_i = 1$)
- $K$ is the strike price
- $n$ is the number of underlying assets in the basket

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)} \max\left( K - \sum_{i=1}^n w_i S_T^i, 0 \right) \middle| \mathcal{F}_t\right]$$

**Note:** Basket put options provide portfolio-level protection against basket depreciation. The option's value exhibits complex dependence on the correlation structure among the basket components - higher correlation increases the basket's overall volatility, making the put option more valuable as simultaneous declines are more likely. The weighted sum $\sum_{i=1}^n w_i S_T^i$ represents the basket value, and the option provides insurance when this diversified portfolio falls below the strike level. Basket puts are typically cheaper than buying individual put options on each component due to the diversification benefit embedded in the basket construction.

### 4.18 Palladium (Optional Dispersion Trade)

**Portfolio Payoff at Expiry:**
$$V_T = \sum_{i=1}^n w_i \cdot \max(\phi(S_T^i - K_i), 0) - \max(\phi(B_T - K_I), 0)$$
where:
- $\phi = 1$ for calls, $\phi = -1$ for puts
- $S_T^i$ is the price of individual asset $i$ at expiry
- $B_T = \sum_{i=1}^n w_i S_T^i$ is the basket/index value at expiry
- $K_i$ is the strike price for individual asset $i$
- $K_I$ is the strike price for the index
- $w_i$ is the weight of asset $i$ in the index

**Approximate P&L Decomposition:**
$$P\&L \approx \left( \sum_{i=1}^n w_i \cdot \sigma_i^2 \right) - \sigma_I^2$$
where:
- $\sigma_i^2$ is the realized variance of asset $i$ over the trade horizon
- $\sigma_I^2$ is the realized variance of the index

**Potential Future Exposure (PFE) at confidence level $\alpha$:**
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where 
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)} \left( \sum_{i=1}^n w_i \cdot \max(\phi(S_T^i - K_i), 0) - \max(\phi(B_T - K_I), 0) \right) \middle| \mathcal{F}_t\right]$$

**Note:** The Palladium trade is a dispersion strategy that expresses a view on future realized correlation. The trade profits when the weighted sum of individual stock variances exceeds the index variance, which occurs when correlation among components is low. This creates negative exposure to correlation - as correlation decreases, dispersion increases and the trade becomes more profitable. The strategy requires sophisticated modeling of the multivariate dynamics and correlation structure of all components in the basket.

## 5. Generalized Potential Future Exposure with Collateralization

### 5.1 Uncollateralized PFE (Recap)

For a derivative with value process $(V_t)_{t \in [0,T]}$, the uncollateralized Potential Future Exposure at confidence level $\alpha$ and time $t$ is defined as:

$$\text{PFE}_\alpha^{\text{uncoll}}(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t^+ \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$

where $V_t^+ = \max(V_t, 0)$ represents the exposure (only positive values create counterparty risk).

### 5.2 Collateralized PFE Framework

Let $C_t$ be the collateral posted at time $t$. The collateralized exposure is:

$$E_t^{\text{coll}} = \max(V_t - C_t, 0)$$

The collateral process is typically governed by a Credit Support Annex (CSA) with:

**Definition 5.2.1** (Collateral Process):
$$C_t = \begin{cases}
V_{t-\delta} & \text{for symmetric collateral} \\
\max(V_{t-\delta}, 0) & \text{for one-way collateral (only posted by counterparty)}
\end{cases}$$

where $\delta$ is the margin period of risk.

### 5.3 Margin Period of Risk (MPoR)

The Margin Period of Risk $\delta$ represents the time between the last collateral exchange and the close-out of positions upon default. It accounts for:

- **Margin Frequency**: Time between regular margin calls
- **Cure Period**: Grace period for posting collateral
- **Close-out Time**: Time to liquidate positions

**Definition 5.3.1** (Effective Exposure with MPoR):
$$E_t^{\text{MPoR}} = \max\left(V_t - C_{t-\delta}, 0\right)$$
where $C_{t-\delta}$ is the last known collateral amount before the default period.

### 5.4 Collateralized PFE Definition

**Definition 5.4.1** (Collateralized PFE):
For a given margin period of risk $\delta > 0$ and confidence level $\alpha$:

$$\text{PFE}_\alpha^{\text{coll}}(t; \delta) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(E_t^{\text{MPoR}} \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$

where:
$$E_t^{\text{MPoR}} = \max\left(V_t - C_{t-\delta}, 0\right)$$

### 5.5 Thresholds and Minimum Transfer Amounts

Real-world collateral agreements include:

- **Threshold** $\theta$: Exposure level below which no collateral is required
- **Minimum Transfer Amount** $MTA$: Minimum collateral transfer amount

**Definition 5.5.1** (Collateral with Thresholds):
$$C_t = \begin{cases}
\max(V_{t-\delta} - \theta, 0) & \text{if } |V_{t-\delta} - C_{t-\delta}| \geq MTA \\
C_{t-\delta} & \text{otherwise}
\end{cases}$$

### 5.6 Practical Computation

For computational purposes, the collateralized PFE can be expressed as:

**Theorem 5.6.1** (Collateralized PFE Representation):
$$\text{PFE}_\alpha^{\text{coll}}(t; \delta) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(V_t - V_{t-\delta} \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$

*Proof:* Under daily margining with zero thresholds, $C_t = V_{t-1}$, so:
$$E_t^{\text{MPoR}} = \max(V_t - V_{t-\delta}, 0)$$

The distribution of $V_t - V_{t-\delta}$ represents the price change over the margin period of risk.

### 5.7 Application to Exotic Derivatives

For any exotic derivative with payoff $\Phi(X)$ and value process:
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\Phi(X) \middle| \mathcal{F}_t\right]$$

The collateralized PFE requires modeling the joint distribution of $(V_t, V_{t-\delta})$, which depends on the path-dependent nature of the exotic payoff and the correlation structure over time intervals of length $\delta$.

## Mathematical Foundations of Exotic Equity Derivatives Pricing

### 1. Basic Probability Framework

Let $(\Omega, \mathcal{F}, \mathbb{P})$ be a complete probability space, where:
- $\Omega$ is the sample space representing all possible market scenarios
- $\mathcal{F}$ is a $\sigma$-algebra representing the set of observable events
- $\mathbb{P}$ is the risk-neutral probability measure

Let $\mathbb{F} = (\mathcal{F}_t)_{t \in [0,T]}$ be a filtration satisfying the usual conditions, representing the flow of information over time interval $[0,T]$.

**Definition 1.1 (Price Process Space)**
The underlying asset price is modeled as a stochastic process:
$$S: [0,T] \times \Omega \to \mathbb{R}^+$$
such that for each $t \in [0,T]$, $S_t$ is $\mathcal{F}_t$-measurable and $S_t$ is a positive semimartingale.

**Definition 1.2 (Path Space)**
Define the path space as:
$$\mathcal{P} = C([0,T], \mathbb{R}^+)$$
the space of continuous functions from $[0,T]$ to $\mathbb{R}^+$. Let $X: \Omega \to \mathcal{P}$ be the path mapping defined by:
$$X(\omega) = (S_t(\omega))_{t \in [0,T]}$$

**Definition 1.3 (Payoff Functional)**
An exotic equity derivative is characterized by its payoff functional:
$$\Phi: \mathcal{P} \to \mathbb{R}^+$$
which is a $\mathcal{B}(\mathcal{P})/\mathcal{B}(\mathbb{R}^+)$-measurable function, where $\mathcal{B}(\cdot)$ denotes the Borel $\sigma$-algebra.

The terminal payoff at maturity $T$ is given by:
$$V_T = \Phi(X)$$

### 2. Fundamental Path Functionals

**Definition 2.1 (Running Maximum)**
For any path $X \in \mathcal{P}$, define the running maximum functional:
$$M_t(X) = \sup\{S_\tau : \tau \in [0,t]\} = \max_{0 \leq \tau \leq t} S_\tau$$

**Definition 2.2 (Running Minimum)**
For any path $X \in \mathcal{P}$, define the running minimum functional:
$$m_t(X) = \inf\{S_\tau : \tau \in [0,t]\} = \min_{0 \leq \tau \leq t} S_\tau$$

**Definition 2.3 (Indicator Functional)**
For any Borel set $A \subset \mathbb{R}^+$, define the indicator functional:
$$\mathbf{1}_A(X_t) = \begin{cases}
1 & \text{if } S_t \in A \\
0 & \text{otherwise}
\end{cases}$$

**Definition 2.4 (Barrier Hit Time)**
For a barrier level $B > 0$, define the hitting time:
$$\tau_B(X) = \inf\{t \in [0,T] : S_t \leq B\} \quad \text{(down-barrier)}$$
$$\tau_B(X) = \inf\{t \in [0,T] : S_t \geq B\} \quad \text{(up-barrier)}$$
The barrier hit indicator is given by:
$$\mathbf{1}_{\{\tau_B \leq T\}}(X) = \begin{cases}
1 & \text{if } \tau_B(X) \leq T \\
0 & \text{otherwise}
\end{cases}$$

### 3. Valuation Theory

**Theorem 3.1 (Risk-Neutral Pricing)**
The no-arbitrage price of the derivative at time $t$ is given by:
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\Phi(X) \middle| \mathcal{F}_t\right]$$
where $r$ is the risk-free rate and $\mathbb{Q}$ is the risk-neutral measure.

**Proof:** By the Fundamental Theorem of Asset Pricing, the absence of arbitrage implies the existence of a risk-neutral measure $\mathbb{Q}$ such that discounted asset prices are martingales. The discounted derivative price process $e^{-rt}V_t$ must be a $\mathbb{Q}$-martingale, hence:
$$e^{-rt}V_t = \mathbb{E}^{\mathbb{Q}}[e^{-rT}V_T | \mathcal{F}_t] = \mathbb{E}^{\mathbb{Q}}[e^{-rT}\Phi(X) | \mathcal{F}_t]$$

### 4. Exposure Measures

**Definition 4.1 (Uncollateralized Potential Future Exposure)**
For a derivative with value process $(V_t)_{t \in [0,T]}$, the uncollateralized potential future exposure (PFE) at time $t$ for confidence level $\alpha$ is defined as:
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{P}\left(\max_{t \leq \tau \leq T} V_\tau^+ \leq x \middle| \mathcal{F}_t\right) \geq \alpha\right\}$$
where $V_\tau^+ = \max(V_\tau, 0)$ represents the positive exposure.

**Definition 4.2 (Collateralized Potential Future Exposure)**
Given a margin period of risk $\Delta$, the collateralized potential future exposure at time $t$ is defined as:
$$\text{PFE}_\alpha^C(t) = \inf\left\{x \in \mathbb{R} : \mathbb{P}\left(\max_{\tau \in [t, t+\Delta]} \left(V_\tau - C_\tau\right)^+ \leq x \middle| \mathcal{F}_t\right) \geq \alpha\right\}$$
where $C_\tau$ represents the collateral amount at time $\tau$, and we assume the account is fully collateralized after each margin call.

## 5. Specific Payoff Structures and Exposure Measures

### 5.1 Lookback Put Option

**Definition 5.1.1 (Lookback Put Payoff)**
The payoff of a lookback put option at expiry time $T$ is defined as:
$$V_T = \max\left(M_T - S_T, 0\right)$$
where $M_T = \sup\{S_\tau : \tau \in [0,T]\}$ is the running maximum of the underlying asset price over the interval $[0,T]$.

**Theorem 5.1.2 (Lookback Put Valuation)**
The no-arbitrage price at time $t \in [0,T]$ is given by:
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\max\left(M_T - S_T, 0\right) \middle| \mathcal{F}_t\right]$$
where $r$ is the risk-free rate and $\mathbb{Q}$ is the risk-neutral measure.

**Definition 5.1.3 (Uncollateralized Potential Future Exposure)**
For the lookback put option, the uncollateralized potential future exposure at confidence level $\alpha \in (0,1)$ is defined as:
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(\sup_{t \leq \tau \leq T} V_\tau^+ \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where $V_\tau^+ = \max(V_\tau, 0)$ represents the positive exposure at time $\tau$.

**Definition 5.1.4 (Collateralized Potential Future Exposure)**
Given a margin period of risk $\Delta > 0$, the collateralized potential future exposure for the lookback put is:
$$\text{PFE}_\alpha^C(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(\sup_{\tau \in [t, t+\Delta]} \left(V_\tau - C_\tau\right)^+ \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where $C_\tau$ is the collateral amount at time $\tau$, and we assume full collateralization after each margin call.

### 5.2 Reset Put Option

**Definition 5.2.1 (Reset Put Payoff)**
The payoff of a reset put option at expiry time $T$ is defined as:
$$V_T = \max\left(K_T - S_T, 0\right)$$
where $K_T$ is the reset strike price, which is determined at predetermined reset dates $\{t_1, t_2, \dots, t_n\} \subset [0,T]$. Typically, $K_T = \min(K_0, S_{t_1}, S_{t_2}, \dots, S_{t_n})$ where $K_0$ is the initial strike price.

**Definition 5.2.2 (Formal Reset Mechanism)**
Let $\mathcal{R} = \{r_1, r_2, \dots, r_n\}$ be the set of reset times with $0 \leq r_1 < r_2 < \cdots < r_n \leq T$. The reset strike process $(K_t)_{t \in [0,T]}$ is defined recursively:
$$K_t = \begin{cases}
K_0 & \text{for } t \in [0, r_1) \\
\min(K_{r_i^-}, S_{r_i}) & \text{for } t = r_i, i = 1, \dots, n \\
K_{r_i} & \text{for } t \in (r_i, r_{i+1})
\end{cases}$$
The final strike is $K_T = K_{r_n}$.

**Theorem 5.2.3 (Reset Put Valuation)**
The no-arbitrage price at time $t \in [0,T]$ is given by:
$$V_t = \mathbb{E}^{\mathbb{Q}}\left[e^{-r(T-t)}\max\left(K_T - S_T, 0\right) \middle| \mathcal{F}_t\right]$$
where $K_T$ is $\mathcal{F}_{r_n}$-measurable and represents the final reset strike level.

**Definition 5.2.4 (Uncollateralized Potential Future Exposure)**
For the reset put option, the uncollateralized potential future exposure at confidence level $\alpha \in (0,1)$ is defined as:
$$\text{PFE}_\alpha(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(\sup_{t \leq \tau \leq T} V_\tau^+ \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where $V_\tau^+ = \max(V_\tau, 0)$ and $V_\tau$ incorporates the reset mechanism up to time $\tau$.

**Definition 5.2.5 (Collateralized Potential Future Exposure)**
Given a margin period of risk $\Delta > 0$, the collateralized potential future exposure for the reset put is:
$$\text{PFE}_\alpha^C(t) = \inf\left\{x \in \mathbb{R} : \mathbb{Q}\left(\sup_{\tau \in [t, t+\Delta]} \left(V_\tau - C_\tau\right)^+ \leq x \middle| \mathcal{F}_0\right) \geq \alpha\right\}$$
where $C_\tau$ is the collateral amount at time $\tau$, accounting for the discrete nature of reset dates in the collateral calculation.

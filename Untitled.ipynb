```python
import numpy as np
from QuantLib import *
import matplotlib.pyplot as plt
```

```python
# 今日日期和市场参数
today = Date.todaysDate()
Settings.instance().evaluationDate = today
calendar = TARGET()
day_count = Actual365Fixed()

# 市场数据
spot_price = 100
strike_price = 95
barrier_level = 85
volatility = 0.2
risk_free_rate = 0.01
maturity = today + Period(1, Years)

# QuantLib Quote 和 TermStructure
spot = SimpleQuote(spot_price)
rate_curve = YieldTermStructureHandle(FlatForward(today, risk_free_rate, day_count))
vol_curve = BlackVolTermStructureHandle(BlackConstantVol(today, calendar, volatility, day_count))

# Black-Scholes-Merton过程
process = BlackScholesProcess(
    QuoteHandle(spot),
    rate_curve,
    vol_curve
)
```

```python
def generate_path(process, maturity_in_years, steps):
    time_grid = TimeGrid(maturity_in_years, steps)
    
    rng = GaussianRandomSequenceGenerator(
        UniformRandomSequenceGenerator(steps, UniformRandomGenerator())
    )
    
    seq = GaussianPathGenerator(process, time_grid, rng, False)
    sample = seq.next()
    path = sample.value()
    return np.array([path[i] for i in range(len(path))])
```

```python
def is_knocked_out(price_path, barrier):
    return np.any(price_path <= barrier)

def down_and_out_put_payoff(price_path, strike, barrier):
    if is_knocked_out(price_path, barrier):
        return 0.0
    else:
        return max(strike - price_path[-1], 0.0)
```

```python
n_paths = 10000
steps = 252
maturity_in_years = 1.0

exposures = []

for _ in range(n_paths):
    path = generate_path(process, maturity_in_years, steps)
    payoff = down_and_out_put_payoff(path, strike_price, barrier_level)
    exposures.append(payoff)

pfe_95 = np.percentile(exposures, 95)
pfe_99 = np.percentile(exposures, 99)

print(f"PFE at 95%: {pfe_95:.4f}")
print(f"PFE at 99%: {pfe_99:.4f}")
```

```python
import QuantLib as ql
import numpy as np
import matplotlib.pyplot as plt

# 参数设定
spot_price = 100
strike_price = 100
barrier_level = 90
volatility = 0.2
risk_free_rate = 0.01
maturity_in_years = 1.0
steps = 50
n_paths = 1000
dt = maturity_in_years / steps

# 构建 QuantLib 组件
calendar = ql.TARGET()
day_count = ql.Actual365Fixed()
todays_date = ql.Date.todaysDate()
ql.Settings.instance().evaluationDate = todays_date

spot_handle = ql.QuoteHandle(ql.SimpleQuote(spot_price))
flat_ts = ql.YieldTermStructureHandle(ql.FlatForward(todays_date, risk_free_rate, day_count))
flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(todays_date, calendar, volatility, day_count))
bsm_process = ql.BlackScholesProcess(spot_handle, flat_ts, flat_vol_ts)

# 生成路径的函数
def generate_path(process, maturity_in_years, steps):
    time_grid = ql.TimeGrid(maturity_in_years, steps)
    rng = ql.GaussianRandomSequenceGenerator(
        ql.UniformRandomSequenceGenerator(steps, ql.UniformRandomGenerator())
    )
    seq = ql.GaussianPathGenerator(process, time_grid, rng, False)
    sample = seq.next()
    path = sample.value()
    return np.array([path[i] for i in range(len(path))])

# 敲出障碍期权的路径敞口计算
def compute_exposure_path(path, strike, barrier):
    exposure_path = []
    knocked_out = False
    for spot in path:
        if spot <= barrier:
            knocked_out = True
        if knocked_out:
            exposure_path.append(0.0)
        else:
            exposure_path.append(max(strike - spot, 0))  # for put
    return exposure_path

# 模拟所有路径
exposures_over_time = []
for _ in range(n_paths):
    path = generate_path(bsm_process, maturity_in_years, steps)
    exposure_path = compute_exposure_path(path, strike_price, barrier_level)
    exposures_over_time.append(exposure_path)

exposures_over_time = np.array(exposures_over_time)

# 计算 EE 和 PFE
expected_exposure = np.mean(exposures_over_time, axis=0)
pfe_95 = np.percentile(exposures_over_time, 95, axis=0)

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(expected_exposure, label='Expected Exposure (EE)', color='blue', linewidth=2)
plt.plot(pfe_95, label='95% PFE', color='red', linestyle='--', linewidth=2)
plt.xlabel('Time Step')
plt.ylabel('Exposure')
plt.title('Exposure Profile Over Time: Down-and-Out Put Option')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 存储所有 spot price 路径
spot_paths = []

# 同时也重新生成 exposures_over_time
exposures_over_time = []

knockout_steps = []  # 记录每条路径 knock out 的 step，如果没有 knock out，记为 None

for _ in range(n_paths):
    path = generate_path(bsm_process, maturity_in_years, steps)
    spot_paths.append(path)
    
    # 敞口计算
    exposure_path = []
    knocked_out = False
    ko_step = None
    for i, spot in enumerate(path):
        if not knocked_out and spot <= barrier_level:
            knocked_out = True
            ko_step = i
        exposure_path.append(0.0 if knocked_out else max(strike_price - spot, 0))
    
    exposures_over_time.append(exposure_path)
    knockout_steps.append(ko_step)

spot_paths = np.array(spot_paths)
exposures_over_time = np.array(exposures_over_time)

# ✅ 绘制所有路径并高亮 knock-out 点
plt.figure(figsize=(10, 6))
time_grid = np.linspace(0, maturity_in_years, steps + 1)

for i in range(n_paths):
    path = spot_paths[i]
    ko_step = knockout_steps[i]
    
    if ko_step is None:
        plt.plot(time_grid, path, color='lightgray', linewidth=0.7)
    else:
        # 在 knock out 之前用蓝色画出来
        plt.plot(time_grid[:ko_step+1], path[:ko_step+1], color='blue', linewidth=0.7)
        # 从 knock out 点之后用灰色
        plt.plot(time_grid[ko_step:], path[ko_step:], color='gray', linestyle='dotted', linewidth=0.7)
        # 标出 knock-out 点
        plt.scatter(time_grid[ko_step], path[ko_step], color='red', s=10, marker='x')

plt.axhline(barrier_level, color='red', linestyle='--', label=f'Barrier: {barrier_level}')
plt.xlabel('Time (years)')
plt.ylabel('Spot Price')
plt.title('Simulated Spot Price Paths with Barrier Knock-Out')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Step 1: 先画路径与 knock-out 点（与之前一样）
plt.figure(figsize=(12, 7))
time_grid = np.linspace(0, maturity_in_years, steps + 1)

for i in range(min(n_paths, 100)):  # 限制最多画100条路径，避免太密
    path = spot_paths[i]
    ko_step = knockout_steps[i]
    
    if ko_step is None:
        plt.plot(time_grid, path, color='lightgray', linewidth=0.7)
    else:
        plt.plot(time_grid[:ko_step+1], path[:ko_step+1], color='blue', linewidth=0.7)
        plt.plot(time_grid[ko_step:], path[ko_step:], color='gray', linestyle='dotted', linewidth=0.7)
        plt.scatter(time_grid[ko_step], path[ko_step], color='red', s=10, marker='x')

# Step 2: 加障碍线
plt.axhline(barrier_level, color='red', linestyle='--', label=f'Barrier: {barrier_level}')

# Step 3: 添加 EE 和 PFE 曲线
expected_exposure = np.mean(exposures_over_time, axis=0)
pfe_95 = np.percentile(exposures_over_time, 95, axis=0)

plt.plot(time_grid, expected_exposure, color='green', linewidth=2, label='Expected Exposure (EE)')
plt.plot(time_grid, pfe_95, color='orange', linewidth=2, linestyle='--', label='95% PFE')

# Final polish
plt.xlabel('Time (years)')
plt.ylabel('Spot Price / Exposure')
plt.title('Barrier Option Paths with Knock-Out Points, EE and PFE')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

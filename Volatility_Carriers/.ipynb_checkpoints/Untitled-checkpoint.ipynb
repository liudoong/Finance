# 1. Variance Swap

## 1.1 Product Definition

Let $(\Omega, \mathcal{F}, \{\mathcal{F}_t\}_{t \geq 0}, \mathbb{P})$ be a filtered probability 
space satisfying the usual conditions. Let $S = \{S_t\}_{t \in [0,T]}$ be an $\mathcal{F}_t$-adapted 
càdlàg semimartingale representing the underlying price process.

The **quadratic variation** of $\log S$ over $[0,T]$ is defined pathwise as:

$$\langle \log S \rangle_T = \lim_{\|\Pi\| \to 0} \sum_{i} \left( \log \frac{S_{t_{i+1}}}{S_{t_i}} \right)^2$$

In discrete time with daily observations, the **annualized realized variance** is:

$$\hat{\sigma}^2_R = \frac{252}{n} \sum_{i=1}^{n} r_i^2, \qquad r_i := \log \frac{S_{t_i}}{S_{t_{i-1}}}$$

No mean correction is applied, consistent with the theoretical definition of quadratic 
variation where $\mathbb{E}[r_i] = \mu \cdot dt = O(dt)$ is asymptotically negligible 
at daily frequency. This is also the convention adopted in standard variance swap 
term sheets (Demeterfi et al., 1999).

A **Variance Swap** is a forward contract whose payoff at maturity $T$ is:

$$\Pi_T = N_{\text{vega}} \cdot \frac{\hat{\sigma}^2_R - K^2}{2K}$$

where $K \in \mathbb{R}_{>0}$ is the **strike volatility** and $N_{\text{vega}}$ is the 
vega notional. The denominator $2K$ linearizes the payoff via first-order Taylor 
expansion around $\hat{\sigma}_R \approx K$:

$$\hat{\sigma}^2_R - K^2 = (\hat{\sigma}_R + K)(\hat{\sigma}_R - K) \approx 2K(\hat{\sigma}_R - K)$$

yielding a **vega-linear** exposure approximation.

---

## 1.2 PFE Methodology

Let $\mathcal{M}^1_+(\mathbb{R})$ denote the space of Borel probability measures on 
$\mathbb{R}$. The **Potential Future Exposure** at confidence level $\alpha \in (0,1)$ 
over margin period of risk $\delta > 0$ is the $\alpha$-quantile of the pushforward 
measure under the historical measure $\mathbb{P}$:

$$\text{PFE}_\alpha = \inf \left\{ x \in \mathbb{R} : \mathbb{P}(V_{t+\delta} - V_t \leq x) \geq \alpha \right\}$$

where $V_t = N_{\text{vega}} \cdot \frac{(\hat{\sigma}_t \cdot K)^2 - K^2}{2K} \cdot \sqrt{T-t}$ 
is the mark-to-market value at time $t$.

We approximate $\mathbb{P}$ by the **empirical measure** $\hat{\mathbb{P}}_n = \frac{1}{n} 
\sum_{i=1}^{n} \delta_{\omega_i} \xrightarrow{w} \mathbb{P}$ (Glivenko-Cantelli). 
The computation proceeds as the following composition of measurable maps:

**Step 1 — Log-return functional** $\phi: \mathbb{R}^n_{>0} \to \mathbb{R}^{n-1}$:

$$\phi(S)_i = \log S_i - \log S_{i+1}$$

**Step 2 — Realized volatility functional** $\sigma_t$, using the second-moment 
estimator consistent with quadratic variation theory:

$$\sigma_t = \sqrt{252 \cdot \frac{1}{252} \sum_{s=0}^{251} r_{t-s}^2} = \sqrt{252 \cdot \overline{r^2_t}}$$

**Step 3 — Shock measure** $T_\delta$, the relative volatility displacement over MPR:

$$\xi_t = \frac{\sigma_t}{\sigma_{t+\delta}} - 1 \in L^0(\Omega, \mathcal{F}, \hat{\mathbb{P}}_n)$$

**Step 4 — Quantile operator** $Q_\alpha$, the generalized inverse of the empirical 
CDF $\hat{F}_\xi \in BV(\mathbb{R})$:

$$\hat{q}_\alpha = \inf \left\{ x : \hat{F}_\xi(x) \geq \alpha \right\}, \qquad
\hat{q}_\alpha = Q_{1-\alpha}[\xi] \text{ for Short}$$

**Step 5 — Exposure map** $\mathcal{E}$, the vega-linear payoff functional:

$$\mathcal{E}(\hat{\sigma}^*) = \frac{(\hat{\sigma}^* \cdot 100)^2 - (K \cdot 100)^2}{2 \cdot K \cdot 100}, 
\qquad \hat{\sigma}^* = K(1 + \hat{q}_\alpha)$$

**Step 6 — Tenor scaling** by Brownian scaling of quadratic variation:

$$\boxed{\text{PFE} = N_{\text{vega}} \cdot \mathcal{E}(\hat{\sigma}^*) \cdot \sqrt{T} 
= N_{\text{vega}} \cdot \sqrt{T} \cdot \mathcal{E} \circ Q_\alpha \circ T_\delta \circ \sigma \circ \phi(S)}$$

```python
import numpy as np
import pandas as pd

# ── Simulation Parameters ──────────────────────────────────────────────
np.random.seed(42)

n_days   = 4 * 252 + 10          # 4 years + 10 trading days = 1018 days
dt       = 1 / 252

# SPX parameters
spx_S0   = 5000.0
spx_mu   = 0.08                  # annual drift
spx_vol  = 0.18                  # annual volatility

# NDX parameters
ndx_S0   = 18000.0
ndx_mu   = 0.10
ndx_vol  = 0.22

rho      = 0.85                  # SPX-NDX correlation

# ── Correlated Brownian Motions ────────────────────────────────────────
Z1 = np.random.standard_normal(n_days)
Z2 = np.random.standard_normal(n_days)
W_spx = Z1
W_ndx = rho * Z1 + np.sqrt(1 - rho**2) * Z2

# ── Simulate Log-Normal Price Paths ───────────────────────────────────
spx_returns = (spx_mu - 0.5 * spx_vol**2) * dt + spx_vol * np.sqrt(dt) * W_spx
ndx_returns = (ndx_mu - 0.5 * ndx_vol**2) * dt + ndx_vol * np.sqrt(dt) * W_ndx

spx_prices = spx_S0 * np.exp(np.cumsum(spx_returns))
ndx_prices = ndx_S0 * np.exp(np.cumsum(ndx_returns))

# ── Build Date Index (most recent first) ──────────────────────────────
end_date   = pd.Timestamp("today").normalize()
dates      = pd.bdate_range(end=end_date, periods=n_days)   # business days

SPX = pd.Series(spx_prices[::-1], index=dates[::-1], name="SPX")
NDX = pd.Series(ndx_prices[::-1], index=dates[::-1], name="NDX")

print(f"SPX: {len(SPX)} observations | Latest: {SPX.iloc[0]:.2f} | Oldest: {SPX.iloc[-1]:.2f}")
print(f"NDX: {len(NDX)} observations | Latest: {NDX.iloc[0]:.2f} | Oldest: {NDX.iloc[-1]:.2f}")
print(f"Date range: {SPX.index[-1].date()} → {SPX.index[0].date()}")
SPX.head()
```

```python
def variance_swap_pfe(underlying_price, tenor, mpr, quantile, N_vega, strike, position):
    """
    Compute Potential Future Exposure (PFE) for Variance Swap.

    Realized variance follows market convention (Demeterfi et al., 1999):

        σ²_R = (252/n) * Σ r_i²,   r_i = log(S_i / S_{i-1})

    No mean correction (Bessel) is applied, consistent with the theoretical
    definition of quadratic variation, where E[r_i] = μ·dt = O(dt) is
    asymptotically negligible at daily frequency.

    The PFE is computed as the α-quantile of the empirical distribution of
    relative volatility shocks over the margin period of risk (MPR), mapped
    through the vega-linear payoff approximation and scaled by √T:

        PFE = N_vega · E(σ*) · √T

    where σ* = K·(1 + q_α) is the shocked volatility and

        E(σ*) = [(σ*·100)² - (K·100)²] / (2·K·100)

    is the vega-linear exposure functional.

    Parameters
    ----------
    underlying_price : pd.Series
        Series of underlying asset prices, with the most recent observation
        on the first row and the oldest observation on the last row.
        Index should be business dates in descending order.
    tenor : float
        Time to maturity in years, e.g., 0.5 for 6 months.
        Used for Brownian scaling of residual exposure: scaling factor = √T.
    mpr : int
        Margin period of risk in trading days.
        Defines the horizon over which the volatility shock is measured:
        ξ_t = σ_t / σ_{t+δ} - 1,  where δ = mpr.
    quantile : float
        Confidence level for PFE, e.g., 0.99 for 99%.
        For Long: q = quantile(ξ, α).
        For Short: q = quantile(ξ, 1-α), reflecting sign reversal of exposure.
    N_vega : float
        Vega notional. Scales the final PFE linearly.
    strike : float
        Strike volatility in decimal, e.g., 0.2 for 20%.
        Enters both the payoff functional E(σ*) and the shocked vol σ* = K(1+q).
    position : str
        "Long"  → PFE uses upper tail quantile q_α  (adverse = vol spike)
        "Short" → PFE uses lower tail quantile q_{1-α} (adverse = vol collapse)

    Returns
    -------
    float
        PFE value in currency units, scaled by N_vega.
        Represents the worst expected loss at the given confidence level
        over the MPR horizon, adjusted for residual tenor.

    Notes
    -----
    - Rolling window is fixed at 252 trading days (one calendar year).
    - Realized vol uses second-moment estimator: σ_t = sqrt(252 * mean(r²)),
      consistent with quadratic variation theory. No Bessel correction applied.
    - Shock series ξ_t is computed in descending time order (most recent first),
      consistent with the input convention of underlying_price.
    - The exposure functional E is C∞, strictly convex, and monotone increasing
      on R_{>0}, ensuring well-defined finite exposure for all σ* > 0.
    """

    # ── Step 1: Log-return functional φ(S) ────────────────────────────────────
    # r_i = log(S_i / S_{i-1}), continuous map on the positive cone of C([0,T])
    # Result is in descending time order (most recent first)
    daily_returns = np.log(underlying_price / underlying_price.shift(-1)).dropna()

    # ── Step 2: Realized volatility functional σ_t ────────────────────────────
    # Market convention: σ_t = sqrt(252 * mean(r²)) over 252-day rolling window
    # Equivalent to annualized quadratic variation rate; no mean or Bessel correction.
    # [::-1] reverses to ascending order for correct rolling direction,
    # then [::-1] restores descending order after computation.
    rolling_realized_vol = np.sqrt(
        daily_returns[::-1].rolling(window=252)
        .apply(lambda x: np.mean(x**2)) * 252
    )
    rolling_realized_vol = rolling_realized_vol[::-1].dropna()

    # ── Step 3: Shock measure T_δ(σ) ─────────────────────────────────────────
    # Relative volatility shock over MPR δ:
    #   ξ_t = σ_t / σ_{t+δ} - 1  ∈ L⁰(Ω, F, P̂_n)
    # Positive ξ_t means vol increased over MPR (adverse for Long position)
    relative_vol_shock = rolling_realized_vol / rolling_realized_vol.shift(-mpr) - 1
    relative_vol_shock = relative_vol_shock.dropna()

    # ── Step 4: Quantile operator Q_α ─────────────────────────────────────────
    # Generalized inverse of empirical CDF F̂_ξ ∈ BV(ℝ) under empirical measure P̂_n
    # Long:  adverse scenario = vol spike    → upper tail α
    # Short: adverse scenario = vol collapse → lower tail 1-α
    if position == "Long":
        quantile_shock_vol = relative_vol_shock.quantile(quantile)
    elif position == "Short":
        quantile_shock_vol = relative_vol_shock.quantile(1 - quantile)
    else:
        raise ValueError("Invalid position: use 'Long' or 'Short'")

    # ── Step 5: Exposure map E(σ*) ────────────────────────────────────────────
    # Shocked volatility: σ* = K · (1 + q_α)
    # Vega-linear payoff approximation (first-order Taylor around σ_R ≈ K):
    #   E(σ*) = [(σ*·100)² - (K·100)²] / (2·K·100)
    # This is C∞, strictly convex, and monotone increasing on ℝ_{>0}
    exposure = ((strike * 100 * (1 + quantile_shock_vol))**2 - (strike * 100)**2) / (2 * strike * 100)

    # ── Step 6: Tenor scaling √T ──────────────────────────────────────────────
    # By Brownian scaling of quadratic variation, residual exposure scales as √T
    exposure = exposure * np.sqrt(tenor)

    # ── Final: Scale by Vega Notional ─────────────────────────────────────────
    # PFE = N_vega · E(σ*) · √T
    return exposure * N_vega
```

```python
# ── Variance Swap PFE ─────────────────────────────────────────────────

# Contract Parameters
tenor    = 0.5        # 6 months
mpr      = 10         # 10 trading days margin period of risk
quantile = 0.99       # 99% quantile
N_vega   = 100_000    # vega notional
strike   = 0.20       # 20% strike volatility
position = "Long"     # Long position

# Compute PFE for SPX and NDX
pfe_spx = variance_swap_pfe(SPX, tenor, mpr, quantile, N_vega, strike, position)
pfe_ndx = variance_swap_pfe(NDX, tenor, mpr, quantile, N_vega, strike, position)

print("=" * 45)
print("        Variance Swap PFE Results")
print("=" * 45)
print(f"  Underlying  : SPX")
print(f"  PFE (99%)   : ${pfe_spx:>12,.2f}")
print("-" * 45)
print(f"  Underlying  : NDX")
print(f"  PFE (99%)   : ${pfe_ndx:>12,.2f}")
print("=" * 45)
print(f"  Parameters  : Tenor={tenor}y | MPR={mpr}d | Strike={strike*100:.0f}% | {position}")
print("=" * 45)
```

# 2. Barrier Variance Swap (Up-Var / Down-Var)

## 2.1 Product Definition

Let $(\Omega, \mathcal{F}, \{\mathcal{F}_t\}_{t \geq 0}, \mathbb{P})$ and $S$ be as defined 
in Section 1. Let $S_0 = S_{t=0}$ denote the initial price at contract inception, and 
let $\beta \in \mathbb{R}_{>0}$ be the **barrier multiplier**. Define the barrier level 
$B := S_0 \cdot \beta$ and the **barrier set**:

$$\mathcal{B} = \begin{cases}
\mathcal{B}^{\text{down}} = \{ x \in \mathbb{R}_{>0} : x < S_0 \cdot \beta \} & \text{Down-Var} \\
\mathcal{B}^{\text{up}} \ \ = \{ x \in \mathbb{R}_{>0} : x > S_0 \cdot \beta \} & \text{Up-Var}
\end{cases}$$

The **accrual indicator process** is $\mathcal{F}_{t-1}$-measurable (one-day lag ensures 
no look-ahead):

$$\mathbf{1}_{\mathcal{B}}(S_{t-1}) \in \{0, 1\}, \qquad t = 1, \ldots, n$$

The **active index set** is $\mathcal{T}_\mathcal{B} := \{ t_i \in \Pi : S_{t_{i-1}} \in \mathcal{B} \}$, 
with accrual fraction $\mu_\mathcal{B} := |\mathcal{T}_\mathcal{B}| / N \in [0,1]$.

The **barrier-conditional realized variance** follows market convention: the denominator 
is always the **total number of trading days** $N$, not $|\mathcal{T}_\mathcal{B}|$:

$$\hat{\sigma}^2_\mathcal{B} = \frac{252}{N} \sum_{t_i \in \mathcal{T}_\mathcal{B}} r_{t_i}^2 
= 252 \cdot \frac{1}{N} \sum_{i=1}^{N} \tilde{r}_i^2$$

where the **barrier-filtered return** $\tilde{r}_i := r_i \cdot \mathbf{1}_\mathcal{B}(S_{t_{i-1}})$ 
assigns zero to inactive days. This is equivalent to:

$$\hat{\sigma}^2_\mathcal{B} = \mu_\mathcal{B} \cdot \hat{\sigma}^2_{R|\mathcal{B}}$$

where $\hat{\sigma}^2_{R|\mathcal{B}}$ is the realized variance conditional on being in the 
barrier regime. Since $\mu_\mathcal{B} \leq 1$, we have the ordering:

$$\hat{\sigma}^2_\mathcal{B} \leq \hat{\sigma}^2_R \qquad \hat{\mathbb{P}}_n\text{-a.s.}$$

The payoff at maturity $T$ is:

$$\Pi_T = N_{\text{vega}} \cdot \frac{\hat{\sigma}^2_\mathcal{B} - K^2}{2K}$$

**Down-Var** ($\beta > 1$, e.g. 115%): variance accrues only when $S_t < S_0 \cdot \beta$, 
capturing left-tail volatility clustering. Typically used to express a view on 
**downside realized variance**.

**Up-Var** ($\beta < 1$, e.g. 80%): variance accrues only when $S_t > S_0 \cdot \beta$, 
isolating upper-regime volatility. Typically used to express a view on 
**upside realized variance**.

---

## 2.2 PFE Methodology

The **Potential Future Exposure** is defined identically to Section 1:

$$\text{PFE}_\alpha = \inf \left\{ x \in \mathbb{R} : \mathbb{P}(V_{t+\delta} - V_t \leq x) \geq \alpha \right\}$$

Under the empirical measure $\hat{\mathbb{P}}_n$, the computation is the following 
composition of measurable maps:

**Step 1 — Log-return functional** $\phi(S)_i = \log S_i - \log S_{i+1}$.

**Step 2 — Barrier restriction operator** $\Pi_\mathcal{B}: L^0(\Omega) \to L^0(\Omega)$:

$$\Pi_\mathcal{B}(r)_i = r_i \cdot \mathbf{1}_\mathcal{B}(S_{t_{i-1}}) =: \tilde{r}_i$$

This is a $\hat{\mathbb{P}}_n$-a.s. well-defined projection on the return process. 
The one-day lag on $S_{t_{i-1}}$ ensures $\mathcal{F}_{t_{i-1}}$-measurability.

**Step 3 — Conditional volatility functional** using full-window denominator:

$$\sigma_{\mathcal{B},t} = \sqrt{252 \cdot \overline{\tilde{r}^2_t}}, \qquad 
\overline{\tilde{r}^2_t} := \frac{1}{252} \sum_{s=0}^{251} \tilde{r}_{t-s}^2$$

**Step 4 — Shock measure**:

$$\xi_t = \frac{\sigma_{\mathcal{B},t}}{\sigma_{\mathcal{B},t+\delta}} - 1 
\in L^0(\Omega, \mathcal{F}, \hat{\mathbb{P}}_n)$$

**Step 5 — Quantile operator** $\hat{q}_\alpha = Q_\alpha[\xi]$ as in Section 1.

**Step 6 — Exposure map** $\mathcal{E}(\hat{\sigma}^*_\mathcal{B})$ as in Section 1, 
with $\hat{\sigma}^*_\mathcal{B} = K(1 + \hat{q}_\alpha)$.

**Step 7 — Tenor scaling**:

$$\boxed{\text{PFE} = N_{\text{vega}} \cdot \sqrt{T} \cdot \mathcal{E} \circ Q_\alpha 
\circ T_\delta \circ \sigma_\mathcal{B} \circ \Pi_\mathcal{B} \circ \phi(S)}$$

Since $\sigma_{\mathcal{B},t} \leq \sigma_t\ \hat{\mathbb{P}}_n$-a.s., it follows that:

$$\text{PFE}_{\text{Barrier}} \leq \text{PFE}_{\text{Vanilla}} \qquad \hat{\mathbb{P}}_n\text{-a.s.}$$

```python
def barrier_variance_swap_pfe(underlying_price, tenor, mpr, quantile, N_vega, strike, position, barrier_type, barrier_value):
    """
    Compute Potential Future Exposure (PFE) for Barrier Variance Swap (Up-Var / Down-Var).

    Realized variance follows market convention: the barrier-inactive days contribute
    zero variance but are still counted in the denominator (total trading days N),
    consistent with standard variance swap term sheets:

        σ²_B = (252 / N_total) * Σ_{t ∈ T_B} r_t²

    where T_B is the active index set and N_total is the total number of trading days
    in the rolling window.

    Down-Var: variance accrues only when S_t < S_0 · β  (β > 1, e.g. 115%)
    Up-Var:   variance accrues only when S_t > S_0 · β  (β < 1, e.g. 80%)

    Parameters
    ----------
    underlying_price : pd.Series
        Price series, most recent first.
    tenor : float
        Time to maturity in years.
    mpr : int
        Margin period of risk in trading days.
    quantile : float
        Confidence level for PFE (e.g., 0.99).
    N_vega : float
        Vega notional.
    strike : float
        Strike volatility in decimal (e.g., 0.2 for 20%).
    position : str
        "Long" or "Short".
    barrier_type : str
        "up" for Up-Var, "down" for Down-Var.
    barrier_value : float
        Barrier level as a fraction of initial price (e.g., 0.8 or 1.15).

    Returns
    -------
    float
        PFE value scaled by N_vega.
    """

    # ── Step 1: Log-return functional φ(S) ────────────────────────────────────
    # r_i = log(S_i / S_{i-1}), descending time order
    daily_returns = np.log(underlying_price / underlying_price.shift(-1)).dropna()

    # ── Step 2: Barrier restriction operator Π_B ──────────────────────────────
    # Barrier reference: initial price at contract inception (most recent observation)
    initial_price = underlying_price.iloc[0]

    # Construct barrier indicator 1_B(S_{t-1}): one-day lag ensures F_{t-1}-measurability
    if barrier_type == "up":
        valid_mask = underlying_price > initial_price * barrier_value
    elif barrier_type == "down":
        valid_mask = underlying_price < initial_price * barrier_value
    else:
        raise ValueError("Invalid barrier_type: use 'up' or 'down'")

    # ── Step 3: Conditional realized volatility σ_{B,t} ───────────────────────
    # Market convention: denominator is N_total (full window = 252),
    # barrier-inactive days contribute r_t = 0 (zero return, not excluded).
    # This correctly penalizes low accrual fraction μ_B = |T_B| / N_total.
    #
    # Implementation: assign zero to barrier-inactive returns, then apply
    # standard second-moment estimator over full 252-day window.
    #
    # σ_{B,t} = sqrt(252 · (1/252) · Σ_{s=0}^{251} r̃²_{t-s})
    #         = sqrt(252 · mean(r̃²))
    # where r̃_i = r_i · 1_B(S_{t_{i-1}})

    # assign zero to inactive days (preserves full window denominator)
    masked_returns = daily_returns.copy()
    masked_returns[~valid_mask.shift(-1).reindex(masked_returns.index).fillna(False)] = 0.0

    # rolling second-moment estimator over full 252-day window
    conditional_vol = np.sqrt(
        masked_returns[::-1].rolling(window=252)
        .apply(lambda x: np.mean(x**2)) * 252
    )
    conditional_vol = conditional_vol[::-1].dropna()

    # ── Step 4: Shock measure T_δ(σ_B) ───────────────────────────────────────
    # ξ_t = σ_{B,t} / σ_{B,t+δ} - 1 ∈ L⁰(Ω, F, P̂_n)
    relative_vol_shock = conditional_vol / conditional_vol.shift(-mpr) - 1
    relative_vol_shock = relative_vol_shock.dropna()

    # ── Step 5: Quantile operator Q_α ─────────────────────────────────────────
    # Generalized inverse of empirical CDF F̂_ξ ∈ BV(ℝ)
    if position == "Long":
        quantile_shock_vol = relative_vol_shock.quantile(quantile)
    elif position == "Short":
        quantile_shock_vol = relative_vol_shock.quantile(1 - quantile)
    else:
        raise ValueError("Invalid position: use 'Long' or 'Short'")

    # ── Step 6: Exposure map E(σ*_B) ──────────────────────────────────────────
    # σ*_B = K · (1 + q_α), vega-linear payoff approximation
    # E(σ*_B) = [(σ*_B · 100)² - (K · 100)²] / (2 · K · 100)
    exposure = ((strike * 100 * (1 + quantile_shock_vol))**2 - (strike * 100)**2) / (2 * strike * 100)

    # ── Step 7: Tenor scaling √T ──────────────────────────────────────────────
    exposure = exposure * np.sqrt(tenor)

    return exposure * N_vega
```

# 3. Corridor Variance Swap

## 3.1 Product Definition

Let $(\Omega, \mathcal{F}, \{\mathcal{F}_t\}_{t \geq 0}, \mathbb{P})$ and $S$ be as defined 
in Section 1. Let $S_0$ denote the initial price, and let $\beta_L, \beta_H \in \mathbb{R}_{>0}$ 
with $\beta_L < \beta_H$ be the **lower and upper barrier multipliers**. Define the 
**corridor** as the open interval:

$$\mathcal{C} := (S_0 \cdot \beta_L,\ S_0 \cdot \beta_H) \subset \mathbb{R}_{>0}$$

This is a Borel-measurable subset of the state space $\mathbb{R}_{>0}$. The 
**corridor indicator process** is $\mathcal{F}_{t-1}$-measurable:

$$\mathbf{1}_\mathcal{C}(S_{t-1}) = \mathbf{1}_{\{S_0 \cdot \beta_L < S_{t-1} < S_0 \cdot \beta_H\}}$$

The **active index set** is $\mathcal{T}_\mathcal{C} := \{t_i \in \Pi : S_{t_{i-1}} \in \mathcal{C}\}$, 
with accrual fraction $\mu_\mathcal{C} := |\mathcal{T}_\mathcal{C}| / N \in [0,1]$.

The **corridor-conditional realized variance** follows market convention with total 
trading days $N$ in the denominator:

$$\hat{\sigma}^2_\mathcal{C} = \frac{252}{N} \sum_{t_i \in \mathcal{T}_\mathcal{C}} r_{t_i}^2 
= 252 \cdot \frac{1}{N} \sum_{i=1}^{N} \tilde{r}_i^2$$

where $\tilde{r}_i := r_i \cdot \mathbf{1}_\mathcal{C}(S_{t_{i-1}})$. The 
corridor variance swap is structurally related to barrier variance swaps via:

$$\hat{\sigma}^2_\mathcal{C} = \hat{\sigma}^2_{\mathcal{B}^{\text{down}}(\beta_H)} 
- \hat{\sigma}^2_{\mathcal{B}^{\text{down}}(\beta_L)}$$

That is, a **Corridor Variance Swap** is the difference of two Down-Var swaps with 
barriers at $\beta_H$ and $\beta_L$ respectively. This decomposition holds 
$\hat{\mathbb{P}}_n$-a.s. and implies the ordering:

$$\hat{\sigma}^2_\mathcal{C} \leq \hat{\sigma}^2_{\mathcal{B}} \leq \hat{\sigma}^2_R 
\qquad \hat{\mathbb{P}}_n\text{-a.s.}$$

The payoff at maturity $T$ is:

$$\Pi_T = N_{\text{vega}} \cdot \frac{\hat{\sigma}^2_\mathcal{C} - K^2}{2K}$$

The corridor structure allows investors to express **regime-specific variance views**: 
by tightening $(\beta_L, \beta_H)$, exposure is concentrated in a narrow price range, 
making the product sensitive to variance accrued exclusively within that regime.

---

## 3.2 PFE Methodology

Under the empirical measure $\hat{\mathbb{P}}_n$, the PFE computation is:

**Step 1 — Log-return functional** $\phi(S)_i = \log S_i - \log S_{i+1}$.

**Step 2 — Corridor restriction operator** $\Pi_\mathcal{C}: L^0(\Omega) \to L^0(\Omega)$:

$$\Pi_\mathcal{C}(r)_i = r_i \cdot \mathbf{1}_\mathcal{C}(S_{t_{i-1}}) =: \tilde{r}_i$$

This is the intersection of two barrier restriction operators:

$$\Pi_\mathcal{C} = \Pi_{\mathcal{B}^{\text{down}}(\beta_H)} \cap \Pi_{\mathcal{B}^{\text{up}}(\beta_L)}$$

i.e. a return is active if and only if the price is simultaneously below $S_0 \cdot \beta_H$ 
and above $S_0 \cdot \beta_L$.

**Step 3 — Conditional volatility functional** with full-window denominator:

$$\sigma_{\mathcal{C},t} = \sqrt{252 \cdot \overline{\tilde{r}^2_t}}, \qquad
\overline{\tilde{r}^2_t} := \frac{1}{252}\sum_{s=0}^{251} \tilde{r}_{t-s}^2$$

**Step 4 — Shock measure**:

$$\xi_t = \frac{\sigma_{\mathcal{C},t}}{\sigma_{\mathcal{C},t+\delta}} - 1 
\in L^0(\Omega, \mathcal{F}, \hat{\mathbb{P}}_n)$$

**Step 5 — Quantile operator** $\hat{q}_\alpha = Q_\alpha[\xi]$ as in Section 1.

**Step 6 — Exposure map** $\mathcal{E}(\hat{\sigma}^*_\mathcal{C})$ as in Section 1, 
with $\hat{\sigma}^*_\mathcal{C} = K(1 + \hat{q}_\alpha)$.

**Step 7 — Tenor scaling**:

$$\boxed{\text{PFE} = N_{\text{vega}} \cdot \sqrt{T} \cdot \mathcal{E} \circ Q_\alpha 
\circ T_\delta \circ \sigma_\mathcal{C} \circ \Pi_\mathcal{C} \circ \phi(S)}$$

The corridor PFE satisfies the chain inequality:

$$\text{PFE}_{\text{Corridor}} \leq \text{PFE}_{\text{Barrier}} \leq \text{PFE}_{\text{Vanilla}} 
\qquad \hat{\mathbb{P}}_n\text{-a.s.}$$

reflecting that each successive restriction of the accrual set reduces the 
expected exposure monotonically.

```python
def corridor_variance_swap_pfe(underlying_price, tenor, mpr, quantile, N_vega, strike, position, barrier_low, barrier_high):
    """
    Compute PFE for Corridor Variance Swap.

    Variance accrues only when the underlying remains within the corridor
    (S_0 · β_low, S_0 · β_high). Consistent with market convention, the
    denominator is the total number of trading days N_total, not the number
    of days within the corridor:

        σ²_corr = (252 / N_total) * Σ_{t ∈ T_corr} r_t²

    Corridor-inactive days contribute zero variance but remain in the denominator,
    penalizing the accrual fraction μ_corr = |T_corr| / N_total ∈ [0,1].

    Parameters
    ----------
    underlying_price : pd.Series
        Price series, most recent first.
    tenor : float
        Time to maturity in years.
    mpr : int
        Margin period of risk in trading days.
    quantile : float
        Confidence level for PFE (e.g., 0.99).
    N_vega : float
        Vega notional.
    strike : float
        Strike volatility in decimal (e.g., 0.2 for 20%).
    position : str
        "Long" or "Short".
    barrier_low : float
        Lower barrier as fraction of initial price (e.g., 0.8 for 80%).
    barrier_high : float
        Upper barrier as fraction of initial price (e.g., 1.2 for 120%).

    Returns
    -------
    float
        PFE value scaled by N_vega.
    """

    # ── Step 1: Log-return functional φ(S) ────────────────────────────────────
    daily_returns = np.log(underlying_price / underlying_price.shift(-1)).dropna()

    # ── Step 2: Corridor restriction operator Π_corr ──────────────────────────
    # Reference price at contract inception
    initial_price = underlying_price.iloc[0]

    # Corridor indicator: 1_{(S_0·β_low, S_0·β_high)}(S_{t-1})
    # One-day lag ensures F_{t-1}-measurability of the accrual indicator
    within_corridor = (
        (underlying_price > initial_price * barrier_low) &
        (underlying_price < initial_price * barrier_high)
    )

    # ── Step 3: Conditional realized volatility σ_{corr,t} ────────────────────
    # Market convention: zero out corridor-inactive returns, keep full denominator.
    # σ_{corr,t} = sqrt(252 · mean(r̃²))
    # where r̃_i = r_i · 1_{corridor}(S_{t_{i-1}})
    masked_returns = daily_returns.copy()
    masked_returns[~within_corridor.shift(-1).reindex(masked_returns.index).fillna(False)] = 0.0

    conditional_vol = np.sqrt(
        masked_returns[::-1].rolling(window=252)
        .apply(lambda x: np.mean(x**2)) * 252
    )
    conditional_vol = conditional_vol[::-1].dropna()

    # ── Step 4: Shock measure T_δ(σ_corr) ────────────────────────────────────
    # ξ_t = σ_{corr,t} / σ_{corr,t+δ} - 1
    relative_vol_shock = conditional_vol / conditional_vol.shift(-mpr) - 1
    relative_vol_shock = relative_vol_shock.dropna()

    # ── Step 5: Quantile operator Q_α ─────────────────────────────────────────
    if position == "Long":
        quantile_shock_vol = relative_vol_shock.quantile(quantile)
    elif position == "Short":
        quantile_shock_vol = relative_vol_shock.quantile(1 - quantile)
    else:
        raise ValueError("Invalid position: use 'Long' or 'Short'")

    # ── Step 6: Exposure map E(σ*_corr) ───────────────────────────────────────
    # σ*_corr = K · (1 + q_α)
    # E(σ*_corr) = [(σ*_corr · 100)² - (K · 100)²] / (2 · K · 100)
    exposure = ((strike * 100 * (1 + quantile_shock_vol))**2 - (strike * 100)**2) / (2 * strike * 100)

    # ── Step 7: Tenor scaling √T ──────────────────────────────────────────────
    exposure = exposure * np.sqrt(tenor)

    return exposure * N_vega
```

# 4. Cross Corridor Variance Swap

## 4.1 Product Definition

Let $(\Omega, \mathcal{F}, \{\mathcal{F}_t\}_{t \geq 0}, \mathbb{P})$ be a filtered probability 
space. Let $S = \{S_t\}_{t \in [0,T]}$ and $A = \{A_t\}_{t \in [0,T]}$ be two 
$\mathcal{F}_t$-adapted càdlàg semimartingales, where $S$ is the **variance underlying** 
(realized variance is computed from $S$) and $A$ is the **observation underlying** 
(barrier condition is evaluated on $A$). Let $A_0$ denote the initial value of $A$ 
at contract inception.

Define the **observation corridor**:

$$\mathcal{C}^A := (A_0 \cdot \beta_L,\ A_0 \cdot \beta_H) \subset \mathbb{R}_{>0}$$

and the $\mathcal{F}_{t-1}$-measurable **cross-corridor indicator**:

$$\mathbf{1}_{\mathcal{C}^A}(A_{t-1}) = \mathbf{1}_{\{A_0 \cdot \beta_L < A_{t-1} < A_0 \cdot \beta_H\}}$$

The **active index set** is $\mathcal{T}_{\mathcal{C}^A} := \{t_i \in \Pi : A_{t_{i-1}} \in \mathcal{C}^A\}$, 
with accrual fraction $\mu_{\mathcal{C}^A} := |\mathcal{T}_{\mathcal{C}^A}| / N \in [0,1]$.

The **cross-corridor realized variance** is computed from $S$ but gated by $A$:

$$\hat{\sigma}^2_{\text{cross}} = \frac{252}{N} \sum_{t_i \in \mathcal{T}_{\mathcal{C}^A}} r^S_{t_i}{}^2 
= 252 \cdot \frac{1}{N} \sum_{i=1}^{N} \tilde{r}^S_i{}^2$$

where $r^S_i := \log(S_{t_i}/S_{t_{i-1}})$ and:

$$\tilde{r}^S_i := r^S_i \cdot \mathbf{1}_{\mathcal{C}^A}(A_{t_{i-1}})$$

This construction decouples the **source of variance** ($S$) from the **accrual 
condition** ($A$), introducing a cross-asset dimension. The joint process 
$(S, A)$ lives on the product space $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$, 
and the cross-corridor realized variance is a functional on the joint path space 
$C([0,T]) \times C([0,T])$.

The payoff at maturity $T$ is:

$$\Pi_T = N_{\text{vega}} \cdot \frac{\hat{\sigma}^2_{\text{cross}} - K^2}{2K}$$

The cross corridor variance swap generalizes the corridor variance swap 
(recovered when $S \equiv A$) and is used to express views on the realized variance 
of one asset **conditional on the price regime of another**, e.g. SPX variance 
accrued only when VIX remains within a given range.

---

## 4.2 PFE Methodology

Under the empirical measure $\hat{\mathbb{P}}_n$, working on the **aligned joint 
date index** $\mathcal{I} = \text{index}(S) \cap \text{index}(A)$, the PFE 
computation proceeds as follows:

**Step 1 — Alignment.** The two series are restricted to the common index:

$$(S, A) : \mathcal{I} \to \mathbb{R}_{>0} \times \mathbb{R}_{>0}$$

ensuring the joint process is well-defined $\hat{\mathbb{P}}_n$-a.s.

**Step 2 — Log-return functional on** $S$:

$$\phi(S)_i = \log S_i - \log S_{i+1} = r^S_i$$

Variance is sourced exclusively from $S$; $A$ contributes only to the accrual indicator.

**Step 3 — Cross-corridor restriction operator** 
$\Pi^{\text{cross}}_{\mathcal{C}^A}: L^0(\Omega) \to L^0(\Omega)$:

$$\Pi^{\text{cross}}_{\mathcal{C}^A}(r^S)_i = r^S_i \cdot \mathbf{1}_{\mathcal{C}^A}(A_{t_{i-1}}) =: \tilde{r}^S_i$$

This operator acts on the return process of $S$ but is conditioned on the path of 
$A$, making it a measurable map on the **product path space** 
$L^0(\Omega_S \times \Omega_A, \mathcal{F}_S \otimes \mathcal{F}_A, \hat{\mathbb{P}}_n)$.

**Step 4 — Conditional volatility functional** with full-window denominator:

$$\sigma^{\text{cross}}_t = \sqrt{252 \cdot \overline{\tilde{r}^S{}^2_t}}, \qquad
\overline{\tilde{r}^S{}^2_t} := \frac{1}{252}\sum_{s=0}^{251} \tilde{r}^S_{t-s}{}^2$$

**Step 5 — Shock measure**:

$$\xi_t = \frac{\sigma^{\text{cross}}_t}{\sigma^{\text{cross}}_{t+\delta}} - 1 
\in L^0(\Omega_S \times \Omega_A, \mathcal{F}_S \otimes \mathcal{F}_A, \hat{\mathbb{P}}_n)$$

**Step 6 — Quantile operator** $\hat{q}_\alpha = Q_\alpha[\xi]$ as in Section 1.

**Step 7 — Exposure map** $\mathcal{E}(\hat{\sigma}^*_{\text{cross}})$ as in Section 1, 
with $\hat{\sigma}^*_{\text{cross}} = K(1 + \hat{q}_\alpha)$.

**Step 8 — Tenor scaling**:

$$\boxed{\text{PFE} = N_{\text{vega}} \cdot \sqrt{T} \cdot \mathcal{E} \circ Q_\alpha 
\circ T_\delta \circ \sigma^{\text{cross}} \circ \Pi^{\text{cross}}_{\mathcal{C}^A} \circ (\phi(S), \mathbf{1}_{\mathcal{C}^A}(A))}$$

When $S \equiv A$, the cross-corridor operator reduces to the standard corridor 
restriction $\Pi_\mathcal{C}$, and the product collapses to a Corridor Variance Swap. 
The general ordering holds:

$$\text{PFE}_{\text{Cross Corridor}} \leq \text{PFE}_{\text{Vanilla}} \qquad \hat{\mathbb{P}}_n\text{-a.s.}$$

```python
def cross_corridor_variance_swap_pfe(underlying_price, observe_price, tenor, mpr, quantile, N_vega, strike, position, barrier_low, barrier_high):
    """
    Compute PFE for Cross Corridor Variance Swap.

    Variance accrual is gated by whether observe_price stays within a corridor
    defined relative to its own initial level, but realized variance is computed
    from underlying_price. Market convention: denominator is N_total (full window),
    not the number of active days:

        σ²_cross = (252 / N_total) * Σ_{t ∈ T_obs} r^{und}_t²

    where T_obs = {t : observe_{t-1} ∈ (obs_0·β_low, obs_0·β_high)}.

    Parameters
    ----------
    underlying_price : pd.Series
        Price series for realized variance computation, most recent first.
    observe_price : pd.Series
        Price series for barrier condition (may be a different asset), most recent first.
    tenor : float
        Time to maturity in years.
    mpr : int
        Margin period of risk in trading days.
    quantile : float
        Confidence level for PFE (e.g., 0.99).
    N_vega : float
        Vega notional.
    strike : float
        Strike volatility in decimal (e.g., 0.2 for 20%).
    position : str
        "Long" or "Short".
    barrier_low : float
        Lower barrier as fraction of initial observe price (e.g., 0.8).
    barrier_high : float
        Upper barrier as fraction of initial observe price (e.g., 1.2).

    Returns
    -------
    float
        PFE value scaled by N_vega.
    """

    # ── Step 0: Align series on common dates ──────────────────────────────────
    underlying_price = underlying_price.squeeze().copy()
    observe_price    = observe_price.squeeze().copy()
    underlying_price, observe_price = underlying_price.align(observe_price, join="inner")

    # ── Step 1: Log-return functional φ(S^{und}) ──────────────────────────────
    # Variance is computed from underlying_price only
    daily_returns = np.log(underlying_price / underlying_price.shift(-1)).dropna()

    # ── Step 2: Cross-corridor restriction operator Π^{cross} ─────────────────
    # Barrier reference: initial observe price at contract inception
    initial_observe = observe_price.iloc[0]

    # Corridor indicator on observe_price: 1_{(obs_0·β_low, obs_0·β_high)}(obs_{t-1})
    # One-day lag: F_{t-1}-measurability of accrual indicator
    corridor_mask = (
        (observe_price > initial_observe * barrier_low) &
        (observe_price < initial_observe * barrier_high)
    )

    # ── Step 3: Conditional realized volatility σ^{cross}_t ───────────────────
    # Zero out underlying returns on days when observe_price is outside corridor.
    # Full 252-day window denominator preserved (market convention).
    # σ^{cross}_t = sqrt(252 · mean(r̃²))
    # where r̃_i = r^{und}_i · 1_{corridor}(obs_{t_{i-1}})
    masked_returns = daily_returns.copy()
    masked_returns[~corridor_mask.shift(-1).reindex(masked_returns.index).fillna(False)] = 0.0

    conditional_vol = np.sqrt(
        masked_returns[::-1].rolling(window=252)
        .apply(lambda x: np.mean(x**2)) * 252
    )
    conditional_vol = conditional_vol[::-1].dropna()

    # ── Step 4: Shock measure T_δ(σ^{cross}) ─────────────────────────────────
    # ξ_t = σ^{cross}_t / σ^{cross}_{t+δ} - 1
    relative_vol_shock = conditional_vol / conditional_vol.shift(-mpr) - 1
    relative_vol_shock = relative_vol_shock.dropna()

    # ── Step 5: Quantile operator Q_α ─────────────────────────────────────────
    if position == "Long":
        quantile_shock_vol = relative_vol_shock.quantile(quantile)
    elif position == "Short":
        quantile_shock_vol = relative_vol_shock.quantile(1 - quantile)
    else:
        raise ValueError("Invalid position: use 'Long' or 'Short'")

    # ── Step 6: Exposure map E(σ*_cross) ──────────────────────────────────────
    # σ*_cross = K · (1 + q_α)
    # E(σ*_cross) = [(σ*_cross · 100)² - (K · 100)²] / (2 · K · 100)
    exposure = ((strike * 100 * (1 + quantile_shock_vol))**2 - (strike * 100)**2) / (2 * strike * 100)

    # ── Step 7: Tenor scaling √T ──────────────────────────────────────────────
    exposure = exposure * np.sqrt(tenor)

    return exposure * N_vega
```

# 5. Cross Variance Contract

## 5.1 Product Definition

Let $(\Omega, \mathcal{F}, \{\mathcal{F}_t\}_{t \geq 0}, \mathbb{P})$ be a filtered 
probability space satisfying the usual conditions. Let $S^1 = \{S^1_t\}_{t \in [0,T]}$ 
and $S^2 = \{S^2_t\}_{t \in [0,T]}$ be two $\mathcal{F}_t$-adapted càdlàg semimartingales. 
Denote their log-returns as $r^k_i := \log(S^k_{t_i} / S^k_{t_{i-1}})$ for $k \in \{1,2\}$.

The **annualized realized variances** are defined via the zero-mean second-moment estimator:

$$\hat{\sigma}^2_k = \frac{252}{N} \sum_{i=1}^{N} (r^k_i)^2 = 252 \cdot \overline{(r^k)^2}, 
\qquad k \in \{1, 2\}$$

The **realized correlation** is defined via the zero-mean cross-moment estimator, 
consistent with quadratic covariation theory:

$$\hat{\rho} = \frac{\frac{1}{N}\sum_{i=1}^{N} r^1_i r^2_i}
{\sqrt{\frac{1}{N}\sum_{i=1}^{N}(r^1_i)^2 \cdot \frac{1}{N}\sum_{i=1}^{N}(r^2_i)^2}}
= \frac{\overline{r^1 r^2}}{\sqrt{\overline{(r^1)^2} \cdot \overline{(r^2)^2}}}$$

This is the discrete analogue of the **quadratic covariation ratio**:

$$\hat{\rho} \approx \frac{[S^1, S^2]_T}{\sqrt{[S^1]_T \cdot [S^2]_T}}$$

No mean correction is applied, since $\mathbb{E}[r^k_i] = \mu_k \cdot dt = O(dt)$ 
is asymptotically negligible at daily frequency.

The **cross variance** is defined as the variance of the return spread 
$r^1 - r^2$, which under the zero-mean assumption decomposes as:

$$\text{CV} := \hat{\sigma}^2_1 + \hat{\sigma}^2_2 - 2\hat{\rho}\hat{\sigma}_1\hat{\sigma}_2 
= \widehat{\text{Var}}(r^1 - r^2) \cdot 252$$

This admits the alternative decomposition:

$$\text{CV} = (\hat{\sigma}_1 - \hat{\sigma}_2)^2 + 2(1 - \hat{\rho})\hat{\sigma}_1\hat{\sigma}_2$$

revealing that cross variance is driven by two components: the **volatility differential** 
$(\hat{\sigma}_1 - \hat{\sigma}_2)^2$ and the **correlation discount** 
$2(1-\hat{\rho})\hat{\sigma}_1\hat{\sigma}_2$. Note that $\text{CV} \geq 0$ 
$\hat{\mathbb{P}}_n$-a.s. by the Cauchy-Schwarz inequality.

The payoff of a **Cross Variance Contract** at maturity $T$ is:

$$\Pi_T = \text{Notional} \cdot (\text{CV} - K)$$

where $K \in \mathbb{R}_{\geq 0}$ is the **strike in variance space** 
(e.g. $K = 0.04$ for a 20% vol level). The payoff is **linear** in cross variance, 
not vega-linearized, distinguishing it from the variance swap family.

---

## 5.2 PFE Methodology

Since cross variance $\text{CV} \geq 0$ is a variance-space quantity and the payoff 
is linear, PFE is computed using an **absolute shock** in variance space (not relative 
as in volatility-space products), with an **option-like floor** at zero.

Under the empirical measure $\hat{\mathbb{P}}_n = \frac{1}{n}\sum_i \delta_{\omega_i}$, 
the computation proceeds as follows:

**Step 1 — Log-return functionals** on the joint path space:

$$\phi(S^k)_i = \log S^k_i - \log S^k_{i+1}, \qquad k \in \{1, 2\}$$

**Step 2 — Variance functionals** $\hat{\sigma}^2_{k,t}$ via zero-mean estimator:

$$\hat{\sigma}^2_{k,t} = 252 \cdot \frac{1}{252}\sum_{s=0}^{251}(r^k_{t-s})^2 
= 252 \cdot \overline{(r^k)^2_t}$$

**Step 3 — Correlation functional** $\hat{\rho}_t$ via quadratic covariation ratio:

$$\hat{\rho}_t = \frac{\overline{r^1 r^2}_t}{\sqrt{\overline{(r^1)^2}_t \cdot \overline{(r^2)^2}_t}}
\in [-1, 1] \qquad \hat{\mathbb{P}}_n\text{-a.s.}$$

by the discrete Cauchy-Schwarz inequality applied under $\hat{\mathbb{P}}_n$.

**Step 4 — Cross variance functional**:

$$\text{CV}_t = \hat{\sigma}^2_{1,t} + \hat{\sigma}^2_{2,t} 
- 2\hat{\rho}_t \hat{\sigma}_{1,t}\hat{\sigma}_{2,t} \geq 0$$

**Step 5 — Absolute shock** $\Delta_t$ in variance space over MPR $\delta$:

$$\Delta_t = \text{CV}_t - \text{CV}_{t+\delta} \in L^0(\Omega, \mathcal{F}, \hat{\mathbb{P}}_n)$$

Absolute (not relative) shock is appropriate since $\text{CV}$ lives in variance 
space $\mathbb{R}_{\geq 0}$, not volatility space $\mathbb{R}_{>0}$.

**Step 6 — Quantile operator** $\hat{q}_\alpha = Q_\alpha[\Delta]$:

$$\hat{q}_\alpha = \inf\{x : \hat{F}_\Delta(x) \geq \alpha\}, \qquad 
\hat{q}_\alpha = Q_{1-\alpha}[\Delta] \text{ for Short}$$

**Step 7 — Shocked cross variance**, floored at zero:

$$\text{CV}^* = \max(\text{CV}_0 + \hat{q}_\alpha,\ 0)$$

**Step 8 — Exposure map with option-like floor**:

$$\mathcal{E}(\text{CV}^*) = \begin{cases}
\max(\text{CV}^* - K,\ 0) & \text{Long} \\
\max(K - \text{CV}^*,\ 0) & \text{Short}
\end{cases}$$

**Step 9 — Tenor scaling**:

$$\boxed{\text{PFE} = \text{Notional} \cdot \mathcal{E}(\text{CV}^*) \cdot \sqrt{T}}$$

The composition of measurable maps is:

$$\text{PFE} = \text{Notional} \cdot \sqrt{T} \cdot \mathcal{E} \circ Q_\alpha \circ \Delta_\delta 
\circ \text{CV} \circ (\hat{\sigma}^2_1, \hat{\sigma}^2_2, \hat{\rho}) \circ (\phi(S^1), \phi(S^2))$$

```python
def cross_variance_contract_pfe(price1, price2, tenor, mpr, quantile, Notional, strike, position):
    """
    Compute PFE of a Cross Variance Contract.

    The cross variance of two assets is defined as the variance of their
    return spread, decomposed via the variance-covariance identity:

        CV = σ1² + σ2² - 2·ρ·σ1·σ2 = (σ1 - σ2)² + 2·(1-ρ)·σ1·σ2

    where σ1, σ2 are annualized realized volatilities and ρ is the realized
    correlation. All three quantities use the zero-mean estimator, consistent
    with quadratic variation theory and market convention:

        σ_i² = (252/N) · Σ r_{i,t}²
        ρ    = Σ r_{1,t}·r_{2,t} / sqrt(Σ r_{1,t}² · Σ r_{2,t}²)

    No mean correction (Bessel) is applied to any estimator.

    The PFE shock is in absolute variance space (not relative), since cross
    variance is a non-negative variance-level quantity:

        Δ_t = CV_t - CV_{t+δ}

    The exposure is floored at zero (option-like), reflecting that PFE
    represents only positive mark-to-market exposure:

        E = max(CV* - K, 0)  for Long
        E = max(K - CV*, 0)  for Short

    where CV* = max(CV_0 + q_α, 0) is the shocked cross variance and
    K is the strike in variance space (e.g. 0.04 for a 20% vol level).

    Parameters
    ----------
    price1, price2 : pd.Series
        Historical price series of the two underlyings, most recent first.
    tenor : float
        Time to maturity in years.
    mpr : int
        Margin period of risk in trading days.
    quantile : float
        Confidence level for PFE (e.g., 0.99).
    Notional : float
        Notional amount.
    strike : float
        Strike in variance space (e.g., 0.04 for a 20% vol strike).
    position : str
        "Long" or "Short".

    Returns
    -------
    float
        PFE value scaled by Notional.
    """

    # ── Step 0: Align series on common dates ──────────────────────────────────
    price1 = price1.squeeze().copy()
    price2 = price2.squeeze().copy()
    price1, price2 = price1.align(price2, join='inner')

    # ── Step 1: Log-return functionals φ(S1), φ(S2) ───────────────────────────
    ret1 = np.log(price1 / price1.shift(-1)).dropna()
    ret2 = np.log(price2 / price2.shift(-1)).dropna()
    ret1, ret2 = ret1.align(ret2, join='inner')

    # ── Step 2: Realized variance functionals σ1², σ2² ────────────────────────
    # Market convention: zero-mean second-moment estimator, no Bessel correction
    # σ_i² = 252 · mean(r_i²)
    var1 = ret1[::-1].rolling(252).apply(lambda x: np.mean(x**2)) * 252
    var2 = ret2[::-1].rolling(252).apply(lambda x: np.mean(x**2)) * 252

    # ── Step 3: Realized correlation functional ρ ──────────────────────────────
    # Zero-mean estimator consistent with quadratic covariation theory:
    # ρ = mean(r1·r2) / sqrt(mean(r1²) · mean(r2²))
    # = [Σ r1·r2] / sqrt([Σ r1²] · [Σ r2²])
    cov12  = ret1[::-1].rolling(252).apply(lambda x: np.mean(x * ret2[::-1].reindex(ret1.index)[ret1[::-1].index]), raw=False)

    # compute element-wise cross product series first, then roll
    cross_prod = ret1 * ret2
    cov12  = cross_prod[::-1].rolling(252).apply(lambda x: np.mean(x))
    denom  = np.sqrt(
        ret1[::-1].rolling(252).apply(lambda x: np.mean(x**2)) *
        ret2[::-1].rolling(252).apply(lambda x: np.mean(x**2))
    )
    rolling_corr = (cov12 / denom).replace([np.inf, -np.inf], np.nan)

    # ── Step 4: Flip and align ────────────────────────────────────────────────
    var1         = var1[::-1].dropna()
    var2         = var2[::-1].dropna()
    rolling_corr = rolling_corr[::-1].dropna()

    var1, var2        = var1.align(var2, join='inner')
    var1, rolling_corr = var1.align(rolling_corr, join='inner')
    var2, rolling_corr = var2.align(rolling_corr, join='inner')

    # ── Step 5: Cross variance functional CV_t ────────────────────────────────
    # CV_t = σ1² + σ2² - 2·ρ·σ1·σ2
    # Equivalent to Var(r1 - r2) under zero-mean assumption
    cross_variance = var1 + var2 - 2 * rolling_corr * np.sqrt(var1) * np.sqrt(var2)
    cross_variance = cross_variance.clip(lower=0)   # CV is non-negative by construction

    # ── Step 6: Absolute shock Δ_t over MPR ───────────────────────────────────
    # Absolute shock in variance space: Δ_t = CV_t - CV_{t+δ}
    # Positive Δ means cross variance increased (adverse for Long)
    shock   = cross_variance - cross_variance.shift(-mpr)
    shock   = shock.dropna()

    # ── Step 7: Quantile operator Q_α ─────────────────────────────────────────
    if position == "Long":
        shock_q = shock.quantile(quantile)
    elif position == "Short":
        shock_q = shock.quantile(1 - quantile)
    else:
        raise ValueError("Invalid position: use 'Long' or 'Short'")

    # ── Step 8: Shocked cross variance CV* ────────────────────────────────────
    # CV* = max(CV_0 + q_α, 0), floored at zero since CV ≥ 0 by construction
    current_cross_var = cross_variance.iloc[0]
    shocked_cross_var = max(current_cross_var + shock_q, 0)

    # ── Step 9: Exposure map with option-like floor ────────────────────────────
    # E = max(CV* - K, 0) for Long   (adverse = CV rises above strike)
    # E = max(K - CV*, 0) for Short  (adverse = CV falls below strike)
    if position == "Long":
        exposure = max(shocked_cross_var - strike, 0)
    else:
        exposure = max(strike - shocked_cross_var, 0)

    # ── Step 10: Tenor scaling √T ─────────────────────────────────────────────
    exposure = exposure * np.sqrt(tenor)

    return exposure * Notional
```

# 6. Correlation Swap

## 6.1 Product Definition

Let $(\Omega, \mathcal{F}, \{\mathcal{F}_t\}_{t \geq 0}, \mathbb{P})$ be a filtered 
probability space satisfying the usual conditions. Let $S^1, S^2$ be two 
$\mathcal{F}_t$-adapted càdlàg semimartingales with log-returns 
$r^k_i := \log(S^k_{t_i}/S^k_{t_{i-1}})$ for $k \in \{1,2\}$.

The **realized correlation** over a window of $n$ trading days is:

$$\hat{\rho}_n = \frac{\frac{1}{n}\sum_{i=1}^{n} r^1_i r^2_i}
{\sqrt{\frac{1}{n}\sum_{i=1}^{n}(r^1_i)^2 \cdot \frac{1}{n}\sum_{i=1}^{n}(r^2_i)^2}}
= \frac{\overline{r^1 r^2}}{\sqrt{\overline{(r^1)^2} \cdot \overline{(r^2)^2}}}$$

This is the ratio of the **empirical quadratic covariation** to the geometric mean of 
the **empirical quadratic variations**, the discrete analogue of:

$$\hat{\rho} \approx \frac{[S^1, S^2]_T}{\sqrt{[S^1]_T \cdot [S^2]_T}} \in [-1,1]$$

where membership in $[-1,1]$ is guaranteed $\hat{\mathbb{P}}_n$-a.s. by the 
Cauchy-Schwarz inequality on $(\mathbb{R}^n, \|\cdot\|_2)$:

$$\left|\sum_i r^1_i r^2_i\right| \leq \sqrt{\sum_i (r^1_i)^2} \cdot \sqrt{\sum_i (r^2_i)^2}$$

No mean correction is applied, since $\mathbb{E}[r^k_i] = O(dt)$ is negligible at 
daily frequency. The window length $n$ is a **contract parameter**, allowing 
correlation to be measured at different horizons (e.g. $n=21$ for one-month 
realized correlation, $n=252$ for one-year).

The payoff of a **Correlation Swap** at maturity $T$ is:

$$\Pi_T = \text{Notional} \cdot (\hat{\rho}_n - K)$$

where $K \in [-1,1]$ is the **strike correlation**. The payoff is **linear** in 
realized correlation. The long holder benefits if realized correlation exceeds the 
strike; the short holder benefits if realized correlation falls below the strike.

Unlike variance swaps, correlation swaps cannot be perfectly replicated by a 
static portfolio of options, making them **genuinely exotic** instruments whose 
pricing requires correlation models (Bossu, 2005). The fair strike is 
approximately:

$$K^* \approx \mathbb{E}^\mathbb{Q}[\hat{\rho}_n] < \bar{\rho}_{\text{implied}}$$

where the inequality reflects the **correlation risk premium**: implied correlation 
systematically exceeds realized correlation, analogous to the variance risk premium.

---

## 6.2 PFE Methodology

Since realized correlation $\hat{\rho} \in [-1,1]$ is a bounded quantity, 
PFE is computed using an **absolute shock** in correlation space, with the 
shocked correlation **clipped** to remain within $[-1,1]$.

Under the empirical measure $\hat{\mathbb{P}}_n = \frac{1}{n}\sum_i \delta_{\omega_i}$:

**Step 1 — Log-return functionals** $\phi(S^k)_i = \log S^k_i - \log S^k_{i+1}$ 
for $k \in \{1,2\}$.

**Step 2 — Realized correlation functional** $\hat{\rho}_t$ via the quadratic 
covariation ratio over rolling window $n$:

$$\hat{\rho}_t = \frac{\overline{r^1 r^2}_t}{\sqrt{\overline{(r^1)^2}_t \cdot \overline{(r^2)^2}_t}}
\in [-1,1] \qquad \hat{\mathbb{P}}_n\text{-a.s.}$$

This defines a measurable map $\rho: L^0(\Omega^2) \to [-1,1]$, where $[-1,1]$ 
is a compact metric space under the Euclidean metric.

**Step 3 — Absolute shock** $\Delta_t$ in correlation space over MPR $\delta$:

$$\Delta_t = \hat{\rho}_t - \hat{\rho}_{t+\delta} \in [-2, 2]$$

Absolute shock is natural here since $\hat{\rho}$ is a bounded, dimensionless 
quantity in $[-1,1]$, unlike volatility which is an unbounded positive quantity.

**Step 4 — Quantile operator** $\hat{q}_\alpha = Q_\alpha[\Delta]$:

$$\hat{q}_\alpha = \inf\{x : \hat{F}_\Delta(x) \geq \alpha\}, \qquad
\hat{q}_\alpha = Q_{1-\alpha}[\Delta] \text{ for Short}$$

**Step 5 — Shocked correlation** $\hat{\rho}^*$, clipped to $[-1,1]$:

$$\hat{\rho}^* = \text{clip}(\hat{\rho}_0 + \hat{q}_\alpha,\ -1,\ 1) 
= \max(-1,\ \min(1,\ \hat{\rho}_0 + \hat{q}_\alpha))$$

The clipping operator $\text{clip}: \mathbb{R} \to [-1,1]$ is the projection 
onto the compact convex set $[-1,1]$, which is a 1-Lipschitz map and hence 
continuous, ensuring measurability of $\hat{\rho}^*$.

**Step 6 — Exposure map with option-like floor**:

$$\mathcal{E}(\hat{\rho}^*) = \begin{cases}
\max(\hat{\rho}^* - K,\ 0) & \text{Long} \\
\max(K - \hat{\rho}^*,\ 0) & \text{Short}
\end{cases}$$

This map is convex and Lipschitz on $[-1,1]$, with Lipschitz constant 1.

**Step 7 — Tenor scaling**:

$$\boxed{\text{PFE} = \text{Notional} \cdot \mathcal{E}(\hat{\rho}^*) \cdot \sqrt{T}}$$

The full composition is:

$$\text{PFE} = \text{Notional} \cdot \sqrt{T} \cdot \mathcal{E} \circ \text{clip}_{[-1,1]} 
\circ Q_\alpha \circ \Delta_\delta \circ \hat{\rho} \circ (\phi(S^1), \phi(S^2))$$

```python
def correlation_swap_pfe(price1, price2, window, mpr, quantile, Notional, strike, position, tenor):
    """
    Compute PFE of a Correlation Swap.

    A correlation swap pays the difference between realized correlation and
    a strike correlation level:

        Payoff = Notional · (ρ_realized - K)

    Realized correlation uses the zero-mean estimator, consistent with
    quadratic covariation theory and market convention:

        ρ = Σ r_{1,t}·r_{2,t} / sqrt(Σ r_{1,t}² · Σ r_{2,t}²)

    No mean correction is applied, as E[r_i] = μ·dt = O(dt) is negligible
    at daily frequency.

    The PFE shock is in absolute correlation space (not relative), since
    correlation is bounded in [-1, 1]:

        Δ_t = ρ_t - ρ_{t+δ}

    The shocked correlation is bounded within [-1, 1] by construction.
    The exposure is floored at zero (option-like):

        E = max(ρ* - K, 0)  for Long
        E = max(K - ρ*, 0)  for Short

    Parameters
    ----------
    price1, price2 : pd.Series
        Historical price series of the two underlyings, most recent first.
    window : int
        Rolling window in trading days (e.g., 252 for 1 year, 21 for 1 month).
    mpr : int
        Margin period of risk in trading days.
    quantile : float
        Confidence level for PFE (e.g., 0.99).
    Notional : float
        Notional amount.
    strike : float
        Strike correlation level in [-1, 1] (e.g., 0.7).
    position : str
        "Long" or "Short".
    tenor : float
        Time to maturity in years.

    Returns
    -------
    float
        PFE value scaled by Notional.
    """

    # ── Step 0: Align series on common dates ──────────────────────────────────
    price1 = price1.squeeze().copy()
    price2 = price2.squeeze().copy()
    price1, price2 = price1.align(price2, join='inner')

    # ── Step 1: Log-return functionals φ(S1), φ(S2) ───────────────────────────
    ret1 = np.log(price1 / price1.shift(-1)).dropna()
    ret2 = np.log(price2 / price2.shift(-1)).dropna()
    ret1, ret2 = ret1.align(ret2, join='inner')

    # ── Step 2: Realized correlation functional ρ_t ───────────────────────────
    # Zero-mean estimator: ρ = mean(r1·r2) / sqrt(mean(r1²) · mean(r2²))
    # Consistent with quadratic covariation: [S1, S2]_T / sqrt([S1]_T · [S2]_T)
    # No Bessel correction, no mean subtraction.
    cross_prod   = ret1 * ret2
    mean_cross   = cross_prod[::-1].rolling(window).apply(lambda x: np.mean(x))
    mean_sq1     = ret1[::-1].rolling(window).apply(lambda x: np.mean(x**2))
    mean_sq2     = ret2[::-1].rolling(window).apply(lambda x: np.mean(x**2))

    rolling_corr = mean_cross / np.sqrt(mean_sq1 * mean_sq2)
    rolling_corr = rolling_corr.replace([np.inf, -np.inf], np.nan)

    # ── Step 3: Flip and drop NaNs ────────────────────────────────────────────
    rolling_corr = rolling_corr[::-1].dropna()

    # ── Step 4: Absolute shock Δ_t over MPR ───────────────────────────────────
    # Absolute shock in correlation space: Δ_t = ρ_t - ρ_{t+δ}
    # Positive Δ = correlation increased (adverse for Long)
    shock = rolling_corr - rolling_corr.shift(-mpr)
    shock = shock.dropna()

    # ── Step 5: Quantile operator Q_α ─────────────────────────────────────────
    if position == "Long":
        shock_q = shock.quantile(quantile)
    elif position == "Short":
        shock_q = shock.quantile(1 - quantile)
    else:
        raise ValueError("Invalid position: use 'Long' or 'Short'")

    # ── Step 6: Shocked correlation ρ*, bounded within [-1, 1] ───────────────
    # ρ* = clip(ρ_0 + q_α, -1, 1)
    # Clipping enforces the natural constraint that correlation ∈ [-1, 1]
    current_corr  = rolling_corr.iloc[0]
    shocked_corr  = np.clip(current_corr + shock_q, -1, 1)

    # ── Step 7: Exposure map with option-like floor ────────────────────────────
    # E = max(ρ* - K, 0)  for Long   (adverse = corr rises above strike)
    # E = max(K - ρ*, 0)  for Short  (adverse = corr falls below strike)
    if position == "Long":
        exposure = max(shocked_corr - strike, 0)
    else:
        exposure = max(strike - shocked_corr, 0)

    # ── Step 8: Tenor scaling √T ──────────────────────────────────────────────
    exposure = exposure * np.sqrt(tenor)

    return exposure * Notional
```
